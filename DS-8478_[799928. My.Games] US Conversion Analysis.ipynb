Comparison by calendar year, regardless of month coverage.

```python
# YoY metrics by calendar year (comparison-only mode)

df_main_my_games_yoy = df_main_my_games.copy()
df_main_my_games_yoy['internal_date'] = pd.to_datetime(df_main_my_games_yoy['internal_date'], errors='coerce')
df_main_my_games_yoy = df_main_my_games_yoy.dropna(subset=['internal_date'])
df_main_my_games_yoy['year'] = df_main_my_games_yoy['internal_date'].dt.year


def _distinct_count(series):
    return series.dropna().nunique()


def _calc_yoy_metrics(group):
    open_cnt = _distinct_count(group['user_access_token'])

    show_sc_cnt = _distinct_count(
        group.loc[group['token_show_shopping_cart'].astype(str) == '1', 'user_access_token']
    )
    interact_cnt = _distinct_count(
        group.loc[group['token_interact'].astype(str) == '1', 'user_access_token']
    )
    create_cnt = _distinct_count(
        group.loc[
            (group['create_invoice_old'] == 1) & (group['instance_id'] == group['bi_instance']),
            'user_access_token'
        ]
    )
    cancel_cnt = _distinct_count(
        group.loc[
            (group['validation_error'] == 1)
            | (group['afs_canceled'] == 1)
            | (group['ps_canceled'] == 1),
            'user_access_token'
        ]
    )
    done_cnt = _distinct_count(
        group.loc[
            (group['instance_id'] == group['bi_instance']) & (group['done_just'] == 1),
            'user_access_token'
        ]
    )

    return pd.Series(
        {
            'Open': open_cnt,
            'CR_OPEN_SHOW_SC': (show_sc_cnt / open_cnt * 100) if open_cnt else np.nan,
            'CR_OPEN_INTERACT': (interact_cnt / open_cnt * 100) if open_cnt else np.nan,
            'CR_OPEN_CREATE': (create_cnt / open_cnt * 100) if open_cnt else np.nan,
            'AFS_PS_Validation_cancel': (cancel_cnt / open_cnt * 100) if open_cnt else np.nan,
            'CR_CREATE_DONE': (done_cnt / create_cnt * 100) if create_cnt else np.nan,
            'CR_OPEN_DONE': (done_cnt / open_cnt * 100) if open_cnt else np.nan,
        }
    )


yoy_metrics_df = (
    df_main_my_games_yoy
    .groupby('year', as_index=False)
    .apply(_calc_yoy_metrics, include_groups=False)
    .reset_index(drop=True)
    .sort_values('year')
)

import sys
from pathlib import Path

_xsolla_root = Path.cwd().parent
if str(_xsolla_root) not in sys.path:
    sys.path.insert(0, str(_xsolla_root))

from html import escape
from IPython.display import HTML, display
from utils.style import XS
from utils.html_style import XS_OUTER_WRAP_START, XS_OUTER_WRAP_END

METRICS_YOY_ORDER = ['CR_OPEN_INTERACT', 'CR_OPEN_CREATE', 'CR_CREATE_DONE', 'CR_OPEN_DONE']
METRICS_YOY_LABELS = {
    'CR_OPEN_INTERACT': 'CR Open-Interact',
    'CR_OPEN_CREATE': 'CR Open-Create',
    'CR_CREATE_DONE': 'CR Create-Done',
    'CR_OPEN_DONE': 'CR Open-Done',
}
year_list_sorted = sorted(yoy_metrics_df['year'].dropna().unique().tolist())

yoy_plot_df = yoy_metrics_df[['year'] + METRICS_YOY_ORDER].copy()

# KPI cards (top row) — Top-3 Locale Metrics style
open_by_year = yoy_metrics_df[['year', 'Open']].copy()
open_kpi_html = ''.join(
    [
        (
            "<div class='kpi-card'>"
            f"<div class='kpi-head'>{int(r['year'])}</div>"
            f"<div class='kpi-value'>{int(r['Open']):,}</div>"
            "<div class='kpi-note'>Unique access tokens</div>"
            "</div>"
        )
        for _, r in open_by_year.iterrows()
    ]
)

# SVG grouped horizontal bars
svg_width = 980
left_pad = 190
right_pad = 90
top_pad = 44
row_gap = 72
bar_h = 18
group_gap = 8
inner_w = svg_width - left_pad - right_pad

max_val = float(np.nanmax(yoy_plot_df[METRICS_YOY_ORDER].to_numpy())) if len(yoy_plot_df) else 1.0
if max_val <= 0:
    max_val = 1.0

year_colors = [getattr(XS, 'S5', XS.PALETTE[0]), getattr(XS, 'S7', XS.PALETTE[1])]

def _xscale(v):
    return (float(v) / max_val) * inner_w if pd.notna(v) else 0.0

rows_svg = []
for row_idx, metric in enumerate(METRICS_YOY_ORDER):
    y_base = top_pad + row_idx * row_gap
    metric_label = METRICS_YOY_LABELS.get(metric, metric)
    rows_svg.append(
        f"<text x='{left_pad - 16}' y='{y_base + bar_h}' text-anchor='end' class='metric-label'>{escape(metric_label)}</text>"
    )
    rows_svg.append(
        f"<line x1='{left_pad}' y1='{y_base + bar_h + 16}' x2='{left_pad + inner_w}' y2='{y_base + bar_h + 16}' class='grid-line' />"
    )

    for y_idx, year in enumerate(year_list_sorted):
        val = float(yoy_plot_df.loc[yoy_plot_df['year'] == year, metric].iloc[0])
        y = y_base + y_idx * (bar_h + group_gap)
        w = _xscale(val)
        color = year_colors[y_idx % len(year_colors)]

        rows_svg.append(
            f"<rect x='{left_pad}' y='{y}' width='{w:.2f}' height='{bar_h}' rx='8' ry='8' fill='{color}' />"
        )
        rows_svg.append(
            f"<text x='{left_pad + w + 8:.2f}' y='{y + bar_h - 4}' class='val-label'>{val:.2f}%</text>"
        )

# X-axis ticks
axis_ticks = []
for t in np.linspace(0, max_val, 6):
    x = left_pad + _xscale(t)
    axis_ticks.append(f"<line x1='{x:.2f}' y1='20' x2='{x:.2f}' y2='{top_pad + len(METRICS_YOY_ORDER)*row_gap - 12}' class='tick-line' />")
    axis_ticks.append(f"<text x='{x:.2f}' y='{top_pad + len(METRICS_YOY_ORDER)*row_gap}' text-anchor='middle' class='tick-label'>{t:.1f}%</text>")

svg_height = top_pad + len(METRICS_YOY_ORDER) * row_gap + 26
legend_html = ''.join(
    [
        f"<div class='legend-item'><span class='sw' style='background:{year_colors[idx % len(year_colors)]};'></span>{int(year)}</div>"
        for idx, year in enumerate(year_list_sorted)
    ]
)

font_body = getattr(XS, 'FONT_BODY', getattr(XS, 'FONT_SANS', "'Roboto', sans-serif"))
font_title = getattr(XS, 'FONT_TITLE', font_body)
font_subtitle = getattr(XS, 'FONT_SUBTITLE', font_body)
color_s10 = getattr(XS, 'S10', '#B2F2FF')
color_s12 = getattr(XS, 'S12', '#EBFCFF')
color_s4 = getattr(XS, 'S4', '#135D6C')
color_s3 = getattr(XS, 'S3', '#0B3841')
color_s7 = getattr(XS, 'S7', '#31CAE8')
grid_major = getattr(XS, 'GRID_MAJOR', '#C0C0C0')
grid_minor = getattr(XS, 'GRID_MINOR', '#E6E6E6')

yoy_html = f"""
<!DOCTYPE html>
<html>
<head>
<meta charset='utf-8'>
<link rel='stylesheet' href='https://fonts.googleapis.com/css2?family=Open+Sans:wght@400;700&family=Roboto:wght@400;500;700&display=swap'>
<style>
  body {{ background:#FFFFFF; margin:0; padding:20px; font-family:{font_body}; color:{XS.TEXT_PRIMARY}; }}
  .chart-wrap {{ background:#FFFFFF; border:none; border-radius:14px; padding:6px 6px 12px; }}
  .title {{ font-size:36px; font-weight:700; margin-bottom:4px; font-family:{font_title}; color:{XS.TEXT_PRIMARY}; line-height:1.1; }}
  .subtitle {{ font-size:14px; color:{XS.TEXT_SECONDARY}; margin-bottom:12px; font-family:{font_subtitle}; }}
  .kpi-row {{ display:grid; grid-template-columns:repeat(2,minmax(0,1fr)); gap:12px; margin-bottom:14px; }}
  .kpi-card {{ border:2px solid #D3D3D3; border-radius:14px; background:#FFFFFF; overflow:hidden; }}
  .kpi-head {{ background:{color_s10}; color:{XS.TEXT_PRIMARY}; font-size:16px; font-weight:700; text-align:center; padding:6px 8px; font-family:{font_title}; }}
  .kpi-value {{ font-size:48px; font-weight:700; color:{color_s7}; margin:8px 0 0; text-align:center; font-family:{font_body}; line-height:1; }}
  .kpi-note {{ font-size:14px; color:{XS.TEXT_SECONDARY}; text-align:center; padding:6px 0 10px; font-family:{font_body}; }}
  .panel {{ background:#FFFFFF; border:2px solid #D0D7DA; border-radius:14px; padding:14px 14px 10px; }}
  .legend {{ display:flex; gap:22px; margin:0 0 10px; }}
  .legend-item {{ font-size:14px; color:{XS.TEXT_SECONDARY}; display:flex; align-items:center; gap:8px; font-family:{font_body}; }}
  .sw {{ width:20px; height:10px; border-radius:5px; display:inline-block; }}
  .metric-label {{ font-size:16px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .val-label {{ font-size:12px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .grid-line {{ stroke:{grid_major}; stroke-width:1; }}
  .tick-line {{ stroke:{grid_minor}; stroke-width:1; }}
  .tick-label {{ font-size:12px; fill:{XS.TEXT_SECONDARY}; font-family:{font_body}; }}
</style>
</head>
<body>
{XS_OUTER_WRAP_START}
<div class='chart-wrap'>
  <div class='title'>YoY Comparison</div>
  <div class='subtitle'>CR Open-Interact · CR Open-Create · CR Create-Done · CR Open-Done</div>
  <div class='kpi-row'>{open_kpi_html}</div>
  <div class='panel'>
    <div class='legend'>{legend_html}</div>
    <svg width='100%' viewBox='0 0 {svg_width} {svg_height}' preserveAspectRatio='xMinYMin meet' role='img' aria-label='YoY comparison chart'>
      {''.join(axis_ticks)}
      {''.join(rows_svg)}
    </svg>
  </div>
</div>
{XS_OUTER_WRAP_END}
</body>
</html>
"""

display(HTML(yoy_html))
```

```python
# YoY comparison by payment method (4 panels in one row)

import sys
from pathlib import Path
from html import escape
from IPython.display import HTML, display

_xsolla_root = Path.cwd().parent
if str(_xsolla_root) not in sys.path:
    sys.path.insert(0, str(_xsolla_root))

from utils.style import XS

# Human-readable metric names
METRICS_YOY_ORDER = ['CR_OPEN_INTERACT', 'CR_OPEN_CREATE', 'CR_CREATE_DONE', 'CR_OPEN_DONE']
METRICS_YOY_LABELS = {
    'CR_OPEN_INTERACT': 'CR Open-Interact',
    'CR_OPEN_CREATE': 'CR Open-Create',
    'CR_CREATE_DONE': 'CR Create-Done',
    'CR_OPEN_DONE': 'CR Open-Done',
}

# Required methods
PAYMENT_METHODS = [
    ('1380.Bank card', 'Bank card'),
    ('3175.Apple Pay', 'Apple Pay'),
    ('24.PayPal', 'PayPal'),
    ('3431.Google Pay', 'Google Pay'),
]

# Base dataframe
df_main_my_games_yoy_all = df_main_my_games.copy()
df_main_my_games_yoy_all['internal_date'] = pd.to_datetime(df_main_my_games_yoy_all['internal_date'], errors='coerce')
df_main_my_games_yoy_all = df_main_my_games_yoy_all.dropna(subset=['internal_date'])
df_main_my_games_yoy_all['year'] = df_main_my_games_yoy_all['internal_date'].dt.year


def _distinct_count(series):
    return series.dropna().nunique()


def _calc_yoy_metrics(group):
    open_cnt = _distinct_count(group['user_access_token'])
    interact_cnt = _distinct_count(
        group.loc[group['token_interact'].astype(str) == '1', 'user_access_token']
    )
    create_cnt = _distinct_count(
        group.loc[
            (group['create_invoice_old'] == 1) & (group['instance_id'] == group['bi_instance']),
            'user_access_token'
        ]
    )
    done_cnt = _distinct_count(
        group.loc[
            (group['instance_id'] == group['bi_instance']) & (group['done_just'] == 1),
            'user_access_token'
        ]
    )

    return pd.Series(
        {
            'Open': open_cnt,
            'CR_OPEN_INTERACT': (interact_cnt / open_cnt * 100) if open_cnt else np.nan,
            'CR_OPEN_CREATE': (create_cnt / open_cnt * 100) if open_cnt else np.nan,
            'CR_CREATE_DONE': (done_cnt / create_cnt * 100) if create_cnt else np.nan,
            'CR_OPEN_DONE': (done_cnt / open_cnt * 100) if open_cnt else np.nan,
        }
    )


font_body = getattr(XS, 'FONT_BODY', getattr(XS, 'FONT_SANS', "'Roboto', sans-serif"))
font_title = getattr(XS, 'FONT_TITLE', font_body)
color_s7 = getattr(XS, 'S7', '#31CAE8')
color_s10 = getattr(XS, 'S10', '#B2F2FF')
panel_bg = '#FFFFFF'
year_colors = [getattr(XS, 'S5', XS.PALETTE[0]), getattr(XS, 'S7', XS.PALETTE[1])]

all_panels_html = []
for pid_value, method_label in PAYMENT_METHODS:
    df_method = df_main_my_games_yoy_all[df_main_my_games_yoy_all['pid'] == pid_value].copy()

    method_metrics_df = (
        df_method
        .groupby('year', as_index=False)
        .apply(_calc_yoy_metrics, include_groups=False)
        .reset_index(drop=True)
        .sort_values('year')
    )

    if method_metrics_df.empty:
        panel_html = f"""
        <div class='method-card'>
          <div class='method-title'>{escape(method_label)}</div>
          <div class='method-sub'>No data available</div>
        </div>
        """
        all_panels_html.append(panel_html)
        continue

    years_sorted = sorted(method_metrics_df['year'].dropna().unique().tolist())

    # Compact per-method SVG (fit inside one card)
    svg_width = 420
    left_pad = 150
    right_pad = 36
    top_pad = 34
    row_gap = 56
    bar_h = 12
    group_gap = 6
    inner_w = svg_width - left_pad - right_pad

    max_val = float(np.nanmax(method_metrics_df[METRICS_YOY_ORDER].to_numpy())) if len(method_metrics_df) else 1.0
    if max_val <= 0:
        max_val = 1.0

    def _xscale(v):
        return (float(v) / max_val) * inner_w if pd.notna(v) else 0.0

    rows_svg = []
    for row_idx, metric in enumerate(METRICS_YOY_ORDER):
        y_base = top_pad + row_idx * row_gap
        metric_label = METRICS_YOY_LABELS.get(metric, metric)
        rows_svg.append(
            f"<text x='{left_pad - 10}' y='{y_base + bar_h}' text-anchor='end' class='metric-label'>{escape(metric_label)}</text>"
        )
        rows_svg.append(
            f"<line x1='{left_pad}' y1='{y_base + bar_h + 12}' x2='{left_pad + inner_w}' y2='{y_base + bar_h + 12}' class='grid-line' />"
        )

        for y_idx, year in enumerate(years_sorted):
            val = float(method_metrics_df.loc[method_metrics_df['year'] == year, metric].iloc[0])
            y = y_base + y_idx * (bar_h + group_gap)
            w = _xscale(val)
            color = year_colors[y_idx % len(year_colors)]

            rows_svg.append(
                f"<rect x='{left_pad}' y='{y}' width='{w:.2f}' height='{bar_h}' rx='6' ry='6' fill='{color}' />"
            )
            rows_svg.append(
                f"<text x='{left_pad + w + 6:.2f}' y='{y + bar_h - 2}' class='val-label'>{val:.1f}%</text>"
            )

    axis_ticks = []
    for t in np.linspace(0, max_val, 5):
        x = left_pad + _xscale(t)
        axis_ticks.append(f"<line x1='{x:.2f}' y1='18' x2='{x:.2f}' y2='{top_pad + len(METRICS_YOY_ORDER)*row_gap - 8}' class='tick-line' />")
        axis_ticks.append(f"<text x='{x:.2f}' y='{top_pad + len(METRICS_YOY_ORDER)*row_gap + 10}' text-anchor='middle' class='tick-label'>{t:.0f}%</text>")

    svg_height = top_pad + len(METRICS_YOY_ORDER) * row_gap + 24

    legend_html = ''.join(
        [
            f"<div class='legend-item'><span class='sw' style='background:{year_colors[idx % len(year_colors)]};'></span>{int(year)}</div>"
            for idx, year in enumerate(years_sorted)
        ]
    )

    open_cards = ''.join(
        [
            (
                "<div class='kpi-chip'>"
                f"<div class='kpi-chip-year'>{int(r['year'])}</div>"
                f"<div class='kpi-chip-val'>{int(r['Open']):,}</div>"
                "</div>"
            )
            for _, r in method_metrics_df[['year', 'Open']].iterrows()
        ]
    )

    panel_html = f"""
    <div class='method-card'>
      <div class='method-title'>{escape(method_label)}</div>
      <div class='method-kpi-row'>{open_cards}</div>
      <div class='method-panel'>
        <div class='legend'>{legend_html}</div>
        <svg width='100%' viewBox='0 0 {svg_width} {svg_height}' preserveAspectRatio='xMinYMin meet' role='img' aria-label='YoY chart for {escape(method_label)}'>
          {''.join(axis_ticks)}
          {''.join(rows_svg)}
        </svg>
      </div>
    </div>
    """
    all_panels_html.append(panel_html)

fig_html = f"""
<!DOCTYPE html>
<html>
<head>
<meta charset='utf-8'>
<link rel='stylesheet' href='https://fonts.googleapis.com/css2?family=Open+Sans:wght@400;700&family=Roboto:wght@400;500;700&display=swap'>
<style>
  body {{ background:#FFFFFF; margin:0; padding:16px; font-family:{font_body}; color:{XS.TEXT_PRIMARY}; }}
  .grid-4 {{ display:grid; grid-template-columns:repeat(4, minmax(0, 1fr)); gap:14px; align-items:start; }}
  .method-card {{ background:#FFFFFF; border:2px solid #D0D7DA; border-radius:14px; padding:10px; }}
  .method-title {{ font-family:{font_title}; font-size:22px; font-weight:700; line-height:1.15; margin:0 0 8px; }}
  .method-kpi-row {{ display:grid; grid-template-columns:repeat(2, minmax(0, 1fr)); gap:8px; margin-bottom:10px; }}
  .kpi-chip {{ border:1px solid #D3D3D3; border-radius:10px; overflow:hidden; background:#FFFFFF; }}
  .kpi-chip-year {{ background:{color_s10}; text-align:center; font-size:12px; font-weight:700; padding:4px 6px; }}
  .kpi-chip-val {{ text-align:center; color:{color_s7}; font-size:26px; font-weight:700; line-height:1.1; padding:6px 4px 8px; }}
  .method-panel {{ background:{panel_bg}; border:1px solid #D0D7DA; border-radius:12px; padding:8px; overflow:hidden; }}
  .legend {{ display:flex; gap:14px; margin:0 0 8px; flex-wrap:wrap; }}
  .legend-item {{ font-size:11px; color:{XS.TEXT_SECONDARY}; display:flex; align-items:center; gap:6px; }}
  .sw {{ width:14px; height:8px; border-radius:4px; display:inline-block; }}
  .metric-label {{ font-size:11px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .val-label {{ font-size:10px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .grid-line {{ stroke:#C0C0C0; stroke-width:1; }}
  .tick-line {{ stroke:#E6E6E6; stroke-width:1; }}
  .tick-label {{ font-size:10px; fill:{XS.TEXT_SECONDARY}; font-family:{font_body}; }}
</style>
</head>
<body>
  <div class='grid-4'>
    {''.join(all_panels_html)}
  </div>
</body>
</html>
"""

display(HTML(fig_html))
```
