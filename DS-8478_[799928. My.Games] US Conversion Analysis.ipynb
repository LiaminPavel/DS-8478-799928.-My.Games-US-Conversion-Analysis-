```python
# YoY metrics by calendar year (comparison-only mode)

billing_yoy_path = '/Users/p.liamin/Documents/Xsolla Researches/DS-8478_[799928. My.Games] US Conversion Analysis/df_main_billing_data_my_games.parquet'
df_main_billing_data_my_games_yoy = pd.read_parquet(billing_yoy_path)
df_main_billing_data_my_games_yoy['created_date'] = pd.to_datetime(df_main_billing_data_my_games_yoy['created_date'], errors='coerce')
df_main_billing_data_my_games_yoy = df_main_billing_data_my_games_yoy.dropna(subset=['created_date'])
df_main_billing_data_my_games_yoy['year'] = df_main_billing_data_my_games_yoy['created_date'].dt.year

yoy_metrics_df = (
    df_main_billing_data_my_games_yoy
    .groupby('year', as_index=False)
    .agg(
        cnt_done_sum=('cnt_done', 'sum'),
        cnt_total_sum=('cnt_total', 'sum'),
    )
    .sort_values('year')
)

yoy_metrics_df['CR_CREATE_DONE'] = np.where(
    yoy_metrics_df['cnt_total_sum'] > 0,
    (yoy_metrics_df['cnt_done_sum'] / yoy_metrics_df['cnt_total_sum']) * 100,
    np.nan,
)
yoy_metrics_df['Open'] = yoy_metrics_df['cnt_done_sum']

import sys
from pathlib import Path

_xsolla_root = Path.cwd().parent
if str(_xsolla_root) not in sys.path:
    sys.path.insert(0, str(_xsolla_root))

from html import escape
from IPython.display import HTML, display, IFrame
from utils.style import XS
from utils.html_style import XS_OUTER_WRAP_START, XS_OUTER_WRAP_END

METRICS_YOY_ORDER = ['CR_CREATE_DONE']
METRICS_YOY_LABELS = {
    'CR_CREATE_DONE': '',
}
year_list_sorted = sorted(yoy_metrics_df['year'].dropna().unique().tolist())

yoy_plot_df = yoy_metrics_df[['year'] + METRICS_YOY_ORDER].copy()

# KPI cards (top row) — Top-3 Locale Metrics style
open_by_year = yoy_metrics_df[['year', 'CR_CREATE_DONE']].copy().sort_values('year')
yoy_sorted_for_diff = yoy_metrics_df.sort_values('year').reset_index(drop=True)


def _yoy_kpi_value_html(year_value: int, cr_value: float) -> str:
    if int(year_value) != 2026:
        return f"<div class='kpi-value'>{cr_value:.2f}%</div>"
    match_idx = yoy_sorted_for_diff.index[yoy_sorted_for_diff['year'] == year_value].tolist()
    if not match_idx:
        return f"<div class='kpi-value'>{cr_value:.2f}%</div>"
    i = match_idx[0]
    if i == 0:
        return f"<div class='kpi-value'>{cr_value:.2f}%</div>"
    prev_cr = yoy_sorted_for_diff.iloc[i - 1]['CR_CREATE_DONE']
    if pd.isna(cr_value) or pd.isna(prev_cr):
        return f"<div class='kpi-value'>{cr_value:.2f}%</div>"
    diff_pp = float(cr_value) - float(prev_cr)
    if diff_pp > 0:
        diff_cls = 'kpi-diff-pos'
        sign = '+'
    elif diff_pp < 0:
        diff_cls = 'kpi-diff-neg'
        sign = ''
    else:
        diff_cls = 'kpi-diff-neu'
        sign = '+'
    diff_txt = f"{sign}{diff_pp:.2f} pp"
    return (
        "<div class='kpi-value'>"
        "<span class='kpi-value-inline'>"
        f"<span class='kpi-value-num'>{cr_value:.2f}%</span>"
        f"<sup class='kpi-diff {diff_cls}'>{diff_txt}</sup>"
        "</span></div>"
    )


open_kpi_html = ''.join(
    [
        (
            "<div class='kpi-card'>"
            f"<div class='kpi-head'>{int(r['year'])}</div>"
            + _yoy_kpi_value_html(int(r['year']), r['CR_CREATE_DONE'])
            + "<div class='kpi-note'>Conversion</div>"
            "</div>"
        )
        for _, r in open_by_year.iterrows()
    ]
)


# SVG grouped horizontal bars
svg_width = 400
left_pad = 32
right_pad = 20
top_pad = 16
bottom_pad = 42
bar_w = 90
group_gap = 20
svg_height = 200
plot_h = svg_height - top_pad - bottom_pad

max_val = float(np.nanmax(yoy_plot_df[METRICS_YOY_ORDER].to_numpy())) if len(yoy_plot_df) else 1.0
if max_val <= 0:
    max_val = 1.0

year_colors = [getattr(XS, 'S5', XS.PALETTE[0]), getattr(XS, 'S7', XS.PALETTE[1])]

def _yscale(v):
    return (float(v) / max_val) * plot_h if pd.notna(v) else 0.0

group_count = len(METRICS_YOY_ORDER)
years_count = max(1, len(year_list_sorted))
group_w = years_count * bar_w + (years_count - 1) * 8
inner_w = svg_width - left_pad - right_pad

group_start_x = left_pad + max(0.0, (inner_w - (group_count * group_w + (group_count - 1) * group_gap)) / 2)

rows_svg = []
axis_ticks = []
for t in np.linspace(0, max_val, 6):
    y = top_pad + (plot_h - (t / max_val) * plot_h)
    axis_ticks.append(f"<line x1='{left_pad}' y1='{y:.2f}' x2='{svg_width - right_pad}' y2='{y:.2f}' class='tick-line' />")

for g_idx, metric in enumerate(METRICS_YOY_ORDER):
    group_x = group_start_x + g_idx * (group_w + group_gap)
    metric_label = METRICS_YOY_LABELS.get(metric, metric)
    if metric_label:
        rows_svg.append(
            f"<text x='{group_x + group_w/2:.2f}' y='{svg_height - 8}' text-anchor='middle' class='metric-label'>{escape(metric_label)}</text>"
        )
    for y_idx, year in enumerate(year_list_sorted):
        val = float(yoy_plot_df.loc[yoy_plot_df['year'] == year, metric].iloc[0])
        h = _yscale(val)
        x = group_x + y_idx * (bar_w + 8)
        y = top_pad + (plot_h - h)
        color = year_colors[y_idx % len(year_colors)]
        rows_svg.append(f"<rect x='{x:.2f}' y='{y:.2f}' width='{bar_w}' height='{h:.2f}' rx='6' ry='6' fill='{color}' />")
        rows_svg.append(f"<text x='{x + bar_w/2:.2f}' y='{y - 6:.2f}' text-anchor='middle' class='val-label'>{val:.2f}%</text>")
        rows_svg.append(f"<text x='{x + bar_w/2:.2f}' y='{top_pad + plot_h + 16:.2f}' text-anchor='middle' class='year-label'>{int(year)}</text>")

legend_html = ''.join(
    [
        f"<div class='legend-item'><span class='sw' style='background:{year_colors[idx % len(year_colors)]};'></span>{int(year)}</div>"
        for idx, year in enumerate(year_list_sorted)
    ]
)

font_body = getattr(XS, 'FONT_BODY', getattr(XS, 'FONT_SANS', "'Roboto', sans-serif"))
font_title = getattr(XS, 'FONT_TITLE', font_body)
font_subtitle = getattr(XS, 'FONT_SUBTITLE', font_body)
color_s10 = getattr(XS, 'S10', '#B2F2FF')
color_s12 = getattr(XS, 'S12', '#EBFCFF')
color_s4 = getattr(XS, 'S4', '#135D6C')
color_s3 = getattr(XS, 'S3', '#0B3841')
color_s7 = getattr(XS, 'S7', '#31CAE8')
grid_major = getattr(XS, 'GRID_MAJOR', '#C0C0C0')
grid_minor = getattr(XS, 'GRID_MINOR', '#E6E6E6')

yoy_html = f"""
<!DOCTYPE html>
<html>
<head>
<meta charset='utf-8'>
<link rel='stylesheet' href='https://fonts.googleapis.com/css2?family=Open+Sans:wght@400;700&family=Roboto:wght@400;500;700&display=swap'>
<script src='https://cdnjs.cloudflare.com/ajax/libs/html2canvas/1.4.1/html2canvas.min.js'></script>
<style>
  body {{ background:#FFFFFF; margin:0; padding:20px; font-family:{font_body}; color:{XS.TEXT_PRIMARY}; }}
  .chart-wrap {{ background:#FFFFFF; border:none; border-radius:14px; padding:6px 6px 12px; max-width: 760px; }}
  .title {{ font-size:36px; font-weight:700; margin-bottom:18px; font-family:{font_title}; color:{XS.TEXT_PRIMARY}; line-height:1.1; }}
  .subtitle {{ font-size:14px; color:{XS.TEXT_SECONDARY}; margin-bottom:12px; font-family:{font_subtitle}; }}
  .kpi-row {{ display:grid; grid-template-columns:repeat(2,minmax(0,1fr)); gap:12px; margin-bottom:14px; }}
  .kpi-card {{ border:2px solid #D3D3D3; border-radius:14px; background:#FFFFFF; overflow:hidden; }}
  .kpi-head {{ background:{color_s10}; color:{XS.TEXT_PRIMARY}; font-size:16px; font-weight:400; text-align:center; padding:6px 8px; font-family:{font_title}; }}
  .kpi-value {{ font-size:48px; font-weight:400; color:{color_s7}; margin:8px 0 0; text-align:center; font-family:{font_body}; line-height:1; display:flex; justify-content:center; align-items:flex-start; }}
  .kpi-value-inline {{ display:inline-flex; align-items:flex-start; gap:4px; }}
  .kpi-value-num {{ font-size:inherit; font-weight:inherit; color:inherit; }}
  .kpi-diff {{ font-size:24px; line-height:1; vertical-align:super; position:relative; top:-0.2em; font-weight:400; margin-left:2px; font-family:{font_body}; }}
  .kpi-diff-pos {{ color:#2E7D32; }}
  .kpi-diff-neg {{ color:#C62828; }}
  .kpi-diff-neu {{ color:#757575; }}
  .kpi-note {{ font-size:14px; color:{XS.TEXT_SECONDARY}; text-align:center; padding:6px 0 10px; font-family:{font_body}; }}
  .panel {{ background:#FFFFFF; border:2px solid #D0D7DA; border-radius:14px; padding:7px 7px 5px; }}
  .legend {{ display:flex; gap:22px; margin:0 0 10px; }}
  .legend-item {{ font-size:14px; color:{XS.TEXT_SECONDARY}; display:flex; align-items:center; gap:8px; font-family:{font_body}; }}
  .sw {{ width:20px; height:10px; border-radius:5px; display:inline-block; }}
  .metric-label {{ font-size:16px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .year-label {{ font-size:12px; fill:{XS.TEXT_SECONDARY}; font-weight:700; font-family:{font_body}; }}
  .val-label {{ font-size:12px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .grid-line {{ stroke:{grid_major}; stroke-width:1; }}
  .tick-line {{ stroke:{grid_minor}; stroke-width:1; }}
  .tick-label {{ font-size:12px; fill:{XS.TEXT_SECONDARY}; font-family:{font_body}; }}
  .yoy-pfw {{ max-width: 760px; margin: 0 auto; }}
  .yoy-save-bottom {{ display:flex; align-items:center; justify-content:flex-end; gap:10px; margin-top:12px; }}
  .yoy-save-btn {{ background:{color_s7}; color:#fff; border:none; border-radius:8px; padding:7px 12px; font-size:12px; font-weight:600; cursor:pointer; }}
  .yoy-save-status {{ font-size:11px; color:{XS.TEXT_SECONDARY}; }}
</style>
</head>
<body>
<div class='yoy-pfw'>
{XS_OUTER_WRAP_START}
<div class='chart-wrap'>
  <div class='title'>YoY Comparison</div>
  <div class='kpi-row'>{open_kpi_html}</div>
  <div class='panel'>
    <svg width='100%' viewBox='0 0 {svg_width} {svg_height}' preserveAspectRatio='xMinYMin meet' role='img' aria-label='YoY comparison chart'>
      {''.join(axis_ticks)}
      {''.join(rows_svg)}
    </svg>
  </div>
</div>
{XS_OUTER_WRAP_END}
  <div class='yoy-save-bottom'><button id='yoy-save-svg' class='yoy-save-btn'>Save as SVG</button><span id='yoy-save-status' class='yoy-save-status'></span></div>
</div>
""" + r"""<script>
// Save full widget as SVG (visual 1:1 using raster snapshot embedded in SVG) — same pattern as Payment Flow
const yoySaveBtn = document.getElementById('yoy-save-svg');
const yoySaveStatus = document.getElementById('yoy-save-status');
if (yoySaveBtn) {
  yoySaveBtn.addEventListener('click', async () => {
    try {
      const root = document.querySelector('.yoy-pfw');
      if (!root) throw new Error('Widget root not found');
      if (typeof html2canvas === 'undefined') throw new Error('html2canvas is not loaded');

      const saveBlock = root.querySelector('.yoy-save-bottom');
      const prevDisplay = saveBlock ? saveBlock.style.display : '';
      if (saveBlock) saveBlock.style.display = 'none';
      const canvas = await html2canvas(root, {
        backgroundColor: '#ffffff',
        useCORS: true,
        scale: 2,
      });
      if (saveBlock) saveBlock.style.display = prevDisplay;

      const pngDataUrl = canvas.toDataURL('image/png');
      const width = canvas.width;
      const height = canvas.height;
      const svgString = `<?xml version="1.0" encoding="UTF-8"?>`
        + `\n<svg xmlns="http://www.w3.org/2000/svg" width="${width}" height="${height}" viewBox="0 0 ${width} ${height}">`
        + `<rect x="0" y="0" width="${width}" height="${height}" fill="white"/>`
        + `<image href="${pngDataUrl}" x="0" y="0" width="${width}" height="${height}"/>`
        + `</svg>`;

      const resp = await fetch('/__save_svg', {
        method: 'POST',
        headers: {'Content-Type': 'image/svg+xml; charset=utf-8'},
        body: svgString,
      });
      if (!resp.ok) throw new Error('Save failed');
      const msg = await resp.text();
      yoySaveStatus.textContent = msg || 'Saved';
      yoySaveStatus.style.color = '#2E7D32';
    } catch (e) {
      yoySaveStatus.textContent = 'Save failed';
      yoySaveStatus.style.color = '#C62828';
    }
  });
}
</script>
""" + """
</body>
</html>
"""

import functools
import http.server
import socketserver
import threading

_yoy_preview_path = (Path.cwd() / "yoy_comparison_preview.html").resolve()
_yoy_preview_path.write_text(yoy_html, encoding="utf-8")


class _YoyTCPServer(socketserver.TCPServer):
    allow_reuse_address = True


class _YoyQuietHandler(http.server.SimpleHTTPRequestHandler):
    def log_message(self, format, *args):
        return

    def do_POST(self):
        if self.path != '/__save_svg':
            self.send_error(404, 'Not Found')
            return
        try:
            content_length = int(self.headers.get('Content-Length', '0'))
            payload = self.rfile.read(content_length)
            target = Path(self.directory) / 'yoy_comparison.svg'
            target.write_bytes(payload)
            self.send_response(200)
            self.send_header('Content-Type', 'text/plain; charset=utf-8')
            self.end_headers()
            self.wfile.write(f'Saved: {target.name}'.encode('utf-8'))
        except Exception as exc:
            self.send_response(500)
            self.send_header('Content-Type', 'text/plain; charset=utf-8')
            self.end_headers()
            self.wfile.write(f'Error: {exc}'.encode('utf-8'))


_yoy_root = _yoy_preview_path.parent
_yoy_httpd = None
_yoy_port = None
for _yoy_try_port in range(8805, 8835):
    try:
        _yoy_handler = functools.partial(
            _YoyQuietHandler,
            directory=str(_yoy_root),
        )
        _yoy_httpd = _YoyTCPServer(("127.0.0.1", _yoy_try_port), _yoy_handler)
        _yoy_port = _yoy_try_port
        break
    except OSError:
        continue

if _yoy_httpd is not None and _yoy_port is not None:
    threading.Thread(target=_yoy_httpd.serve_forever, daemon=True).start()
    _yoy_iframe_url = f"http://127.0.0.1:{_yoy_port}/{_yoy_preview_path.name}"
    print("YoY Comparison embedded chart (HTTP iframe, scripts allowed):", _yoy_iframe_url)
    display(IFrame(src=_yoy_iframe_url, width="100%", height=420))
else:
    print("Could not start local HTTP preview for YoY; open yoy_comparison_preview.html manually.")
    display(HTML(yoy_html))
```

```python
# YoY comparison by payment method (4 panels in one row)

import sys
from pathlib import Path
from html import escape
from IPython.display import HTML, display

_xsolla_root = Path.cwd().parent
if str(_xsolla_root) not in sys.path:
    sys.path.insert(0, str(_xsolla_root))

from utils.style import XS

# Human-readable metric names
METRICS_YOY_ORDER = ['CR_CREATE_DONE']
METRICS_YOY_LABELS = {
    'CR_CREATE_DONE': 'Conversion',
}

# Required methods
PAYMENT_METHODS = [
    (1380, 'Bank card'),
    (3175, 'Apple Pay'),
    (24, 'PayPal'),
    (3431, 'Google Pay'),
]

# Base dataframe (billing source, same as YoY Comparison)
billing_yoy_path = '/Users/p.liamin/Documents/Xsolla Researches/DS-8478_[799928. My.Games] US Conversion Analysis/df_main_billing_data_my_games.parquet'
df_main_billing_data_my_games_yoy_all = pd.read_parquet(billing_yoy_path)
df_main_billing_data_my_games_yoy_all['created_date'] = pd.to_datetime(df_main_billing_data_my_games_yoy_all['created_date'], errors='coerce')
df_main_billing_data_my_games_yoy_all = df_main_billing_data_my_games_yoy_all.dropna(subset=['created_date'])
df_main_billing_data_my_games_yoy_all['year'] = df_main_billing_data_my_games_yoy_all['created_date'].dt.year


font_body = getattr(XS, 'FONT_BODY', getattr(XS, 'FONT_SANS', "'Roboto', sans-serif"))
font_title = getattr(XS, 'FONT_TITLE', font_body)
color_s7 = getattr(XS, 'S7', '#31CAE8')
color_s10 = getattr(XS, 'S10', '#B2F2FF')
panel_bg = '#FFFFFF'
year_colors = [getattr(XS, 'S5', XS.PALETTE[0]), getattr(XS, 'S7', XS.PALETTE[1])]

all_panels_html = []
for instance_id_value, method_label in PAYMENT_METHODS:
    df_method = df_main_billing_data_my_games_yoy_all[
        df_main_billing_data_my_games_yoy_all['instance_id'] == instance_id_value
    ].copy()

    method_metrics_df = (
        df_method
        .groupby('year', as_index=False)
        .agg(
            cnt_done_sum=('cnt_done', 'sum'),
            cnt_total_sum=('cnt_total', 'sum'),
        )
        .sort_values('year')
    )
    method_metrics_df['CR_CREATE_DONE'] = np.where(
        method_metrics_df['cnt_total_sum'] > 0,
        (method_metrics_df['cnt_done_sum'] / method_metrics_df['cnt_total_sum']) * 100,
        np.nan,
    )
    method_metrics_df['Open'] = method_metrics_df['cnt_done_sum']

    if method_metrics_df.empty:
        panel_html = f"""
        <div class='method-card'>
          <div class='method-title'>{escape(method_label)}</div>
          <div class='method-sub'>No data available</div>
        </div>
        """
        all_panels_html.append(panel_html)
        continue

    years_sorted = sorted(method_metrics_df['year'].dropna().unique().tolist())

    # Compact per-method SVG (same logic as YoY Comparison)
    svg_width = 400
    left_pad = 32
    right_pad = 20
    top_pad = 16
    bottom_pad = 42
    bar_w = 90
    group_gap = 20
    svg_height = 200
    plot_h = svg_height - top_pad - bottom_pad

    max_val = float(np.nanmax(method_metrics_df[METRICS_YOY_ORDER].to_numpy())) if len(method_metrics_df) else 1.0
    if max_val <= 0:
        max_val = 1.0

    def _yscale(v):
        return (float(v) / max_val) * plot_h if pd.notna(v) else 0.0

    group_count = len(METRICS_YOY_ORDER)
    years_count = max(1, len(years_sorted))
    group_w = years_count * bar_w + (years_count - 1) * 8
    inner_w = svg_width - left_pad - right_pad
    group_start_x = left_pad + max(0.0, (inner_w - (group_count * group_w + (group_count - 1) * group_gap)) / 2)

    rows_svg = []
    axis_ticks = []
    for t in np.linspace(0, max_val, 6):
        y = top_pad + (plot_h - (t / max_val) * plot_h)
        axis_ticks.append(f"<line x1='{left_pad}' y1='{y:.2f}' x2='{svg_width - right_pad}' y2='{y:.2f}' class='tick-line' />")

    for g_idx, metric in enumerate(METRICS_YOY_ORDER):
        group_x = group_start_x + g_idx * (group_w + group_gap)
        metric_label = METRICS_YOY_LABELS.get(metric, metric)
        rows_svg.append(
            f"<text x='{group_x + group_w/2:.2f}' y='{svg_height - 8}' text-anchor='middle' class='metric-label'>{escape(metric_label)}</text>"
        )
        for y_idx, year in enumerate(years_sorted):
            val = float(method_metrics_df.loc[method_metrics_df['year'] == year, metric].iloc[0])
            h = _yscale(val)
            x = group_x + y_idx * (bar_w + 8)
            y = top_pad + (plot_h - h)
            color = year_colors[y_idx % len(year_colors)]
            rows_svg.append(f"<rect x='{x:.2f}' y='{y:.2f}' width='{bar_w}' height='{h:.2f}' rx='6' ry='6' fill='{color}' />")
            rows_svg.append(f"<text x='{x + bar_w/2:.2f}' y='{y - 6:.2f}' text-anchor='middle' class='val-label'>{val:.2f}%</text>")
            rows_svg.append(f"<text x='{x + bar_w/2:.2f}' y='{top_pad + plot_h + 16:.2f}' text-anchor='middle' class='year-label'>{int(year)}</text>")

    legend_html = ''.join(
        [
            f"<div class='legend-item'><span class='sw' style='background:{year_colors[idx % len(year_colors)]};'></span>{int(year)}</div>"
            for idx, year in enumerate(years_sorted)
        ]
    )

    method_yoy_sorted = method_metrics_df.sort_values('year').reset_index(drop=True)

    def _yoy_method_chip_val_html(year_value: int, cr_value: float) -> str:
        if int(year_value) != 2026:
            return f"<div class='kpi-chip-val'>{float(cr_value):.2f}%</div>"
        match_idx = method_yoy_sorted.index[method_yoy_sorted['year'] == year_value].tolist()
        if not match_idx:
            return f"<div class='kpi-chip-val'>{float(cr_value):.2f}%</div>"
        i = match_idx[0]
        if i == 0:
            return f"<div class='kpi-chip-val'>{float(cr_value):.2f}%</div>"
        prev_cr = method_yoy_sorted.iloc[i - 1]['CR_CREATE_DONE']
        if pd.isna(cr_value) or pd.isna(prev_cr):
            return f"<div class='kpi-chip-val'>{float(cr_value):.2f}%</div>"
        diff_pp = float(cr_value) - float(prev_cr)
        if diff_pp > 0:
            diff_cls = 'kpi-diff-pos'
            sign = '+'
        elif diff_pp < 0:
            diff_cls = 'kpi-diff-neg'
            sign = ''
        else:
            diff_cls = 'kpi-diff-neu'
            sign = '+'
        diff_txt = f"{sign}{diff_pp:.2f} pp"
        return (
            "<div class='kpi-chip-val'>"
            "<span class='kpi-chip-val-inline'>"
            f"<span class='kpi-chip-val-num'>{float(cr_value):.2f}%</span>"
            f"<sup class='kpi-diff {diff_cls}'>{diff_txt}</sup>"
            "</span></div>"
        )

    open_cards = ''.join(
        [
            (
                "<div class='kpi-chip'>"
                f"<div class='kpi-chip-year'>{int(r['year'])}</div>"
                + _yoy_method_chip_val_html(int(r['year']), r['CR_CREATE_DONE'])
                + "</div>"
            )
            for _, r in method_yoy_sorted[['year', 'CR_CREATE_DONE']].iterrows()
        ]
    )

    panel_html = f"""
    <div class='method-card'>
      <div class='method-title'>{escape(method_label)}</div>
      <div class='method-kpi-row'>{open_cards}</div>
      <div class='method-panel'>
        <svg width='100%' viewBox='0 0 {svg_width} {svg_height}' preserveAspectRatio='xMinYMin meet' role='img' aria-label='YoY chart for {escape(method_label)}'>
          {''.join(axis_ticks)}
          {''.join(rows_svg)}
        </svg>
      </div>
    </div>
    """
    all_panels_html.append(panel_html)

fig_html = f"""
<!DOCTYPE html>
<html>
<head>
<meta charset='utf-8'>
<link rel='stylesheet' href='https://fonts.googleapis.com/css2?family=Open+Sans:wght@400;700&family=Roboto:wght@400;500;700&display=swap'>
<style>
  body {{ background:#FFFFFF; margin:0; padding:16px; font-family:{font_body}; color:{XS.TEXT_PRIMARY}; }}
  .main-title {{ font-family:{font_title}; font-size:32px; font-weight:700; line-height:1.1; margin:0 0 12px; }}
  .grid-4 {{ display:grid; grid-template-columns:repeat(4, minmax(0, 1fr)); gap:14px; align-items:start; }}
  .method-card {{ background:#FFFFFF; border:2px solid #D0D7DA; border-radius:14px; padding:10px; }}
  .method-title {{ font-family:{font_title}; font-size:22px; font-weight:700; line-height:1.15; margin:0 0 8px; }}
  .method-kpi-row {{ display:grid; grid-template-columns:repeat(2, minmax(0, 1fr)); gap:8px; margin-bottom:10px; }}
  .kpi-chip {{ border:1px solid #D3D3D3; border-radius:10px; overflow:hidden; background:#FFFFFF; }}
  .kpi-chip-year {{ background:{color_s10}; text-align:center; font-size:12px; font-weight:700; padding:4px 6px; }}
  .kpi-chip-val {{ text-align:center; color:{color_s7}; font-size:26px; font-weight:700; line-height:1.1; padding:6px 4px 8px; display:flex; justify-content:center; align-items:flex-start; font-family:{font_body}; }}
  .kpi-chip-val-inline {{ display:inline-flex; align-items:flex-start; gap:4px; }}
  .kpi-chip-val-num {{ font-size:inherit; font-weight:inherit; color:inherit; }}
  .kpi-chip-val .kpi-diff {{ font-size:13px; line-height:1; vertical-align:super; position:relative; top:-0.2em; font-weight:400; margin-left:2px; font-family:{font_body}; }}
  .kpi-diff-pos {{ color:#2E7D32; }}
  .kpi-diff-neg {{ color:#C62828; }}
  .kpi-diff-neu {{ color:#757575; }}
  .method-panel {{ background:{panel_bg}; border:1px solid #D0D7DA; border-radius:12px; padding:8px; overflow:hidden; }}
  .legend {{ display:flex; gap:14px; margin:0 0 8px; flex-wrap:wrap; }}
  .legend-item {{ font-size:11px; color:{XS.TEXT_SECONDARY}; display:flex; align-items:center; gap:6px; }}
  .sw {{ width:14px; height:8px; border-radius:4px; display:inline-block; }}
  .metric-label {{ font-size:11px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .year-label {{ font-size:10px; fill:{XS.TEXT_SECONDARY}; font-weight:700; font-family:{font_body}; }}
  .val-label {{ font-size:10px; fill:{XS.TEXT_PRIMARY}; font-weight:700; font-family:{font_body}; }}
  .grid-line {{ stroke:#C0C0C0; stroke-width:1; }}
  .tick-line {{ stroke:#E6E6E6; stroke-width:1; }}
  .tick-label {{ font-size:10px; fill:{XS.TEXT_SECONDARY}; font-family:{font_body}; }}
</style>
</head>
<body>
  <div class='main-title'>YoY Comparison by Payment Method</div>
  <div class='grid-4'>
    {''.join(all_panels_html)}
  </div>
</body>
</html>
"""

display(HTML(fig_html))
```

```python
# Payment Flow (df_payment_flow_my_games) — Template 1 structure, Xsolla palette, session step logic

import gc
import json
import sys
from pathlib import Path

import pandas as pd
from IPython.display import HTML, IFrame, display

_cwd = Path.cwd()
_xsolla_root = _cwd if (_cwd / "utils" / "style.py").is_file() else _cwd.parent
if str(_xsolla_root) not in sys.path:
    sys.path.insert(0, str(_xsolla_root))
from utils.style import XS

# =========================
# 1) Parameters
# =========================

PF_MAX_STEPS = 10
PF_TOP_METHODS = 10
PF_DATE_COL = "internal_date"

_xs_pal = [XS.S3, XS.S5, XS.S7, XS.S9, XS.S6, XS.S8, XS.S4, XS.S10]


def pf_pid_label(raw):
    if pd.isna(raw):
        return "Unknown"
    s = str(raw).strip()
    parts = s.split(".", 1)
    return parts[1].strip() if len(parts) > 1 else s


def pf_row_flow_label(row):
    if int(row["sankey_step"]) == 0:
        return pf_pid_label(row["pid"])
    if int(row["done_just"]) == 1:
        bp = row["bill_pid"]
        if pd.notna(bp) and str(bp).strip():
            return pf_pid_label(bp)
    return pf_pid_label(row["pid"])


def pf_build_colors(names):
    out = {"Other": XS.C_OTHER}
    names = sorted(n for n in names if n and n != "Other")
    for i, n in enumerate(names):
        out[n] = _xs_pal[i % len(_xs_pal)]
    return out


# =========================
# 2) Data preparation
# =========================

pf_df = df_payment_flow_my_games.copy()
pf_df = pf_df.dropna(subset=["pid"]).copy()
pf_df["internal_time"] = pd.to_datetime(pf_df["internal_time"], errors="coerce")
pf_df = pf_df.sort_values(["user_access_token", "steps"]).reset_index(drop=True)
pf_df["sankey_step"] = pf_df.groupby("user_access_token", sort=False).cumcount().astype("int64")
pf_df["flow_label"] = pf_df.apply(pf_row_flow_label, axis=1)
pf_df["pid_label"] = pf_df["flow_label"]
gc.collect()

try:
    pf_date_range = f"{pf_df[PF_DATE_COL].min()} – {pf_df[PF_DATE_COL].max()}"
except Exception:
    pf_date_range = ""

pf_df_entry = (
    pf_df.loc[pf_df["sankey_step"] == 0, ["user_access_token", "pid_label"]]
    .drop_duplicates(subset=["user_access_token"])
    .rename(columns={"pid_label": "entry_method"})
    .copy()
)

# Entry step labels: keep top-5 methods, group all others into Other.
pf_entry_counts_all = pf_df_entry["entry_method"].value_counts()
pf_entry_top5 = set(pf_entry_counts_all.head(5).index)
pf_entry_mandatory = {"Bank card", "Apple Pay", "PayPal", "Google Pay"}
pf_entry_allowed = (pf_entry_top5 | pf_entry_mandatory) - {"Unknown"}
pf_df_entry["entry_method"] = pf_df_entry["entry_method"].apply(
    lambda x: x if x in pf_entry_allowed else "Other"
)

pf_entry_counts = pf_df_entry["entry_method"].value_counts()
pf_entry_methods_sorted = pf_entry_counts.index.tolist()
pf_total_users = len(pf_df_entry)

pf_df_after = pf_df.loc[pf_df["sankey_step"] >= 1].copy()
pf_df_after = pf_df_after.merge(
    pf_df_entry[["user_access_token", "entry_method"]], on="user_access_token", how="inner"
)
pf_df_after.sort_values(["user_access_token", "sankey_step"], inplace=True)
pf_df_after.reset_index(drop=True, inplace=True)
pf_df_after["step"] = pf_df_after["sankey_step"].astype("int16")
pf_df_after = pf_df_after.loc[pf_df_after["step"] <= PF_MAX_STEPS].copy()
gc.collect()


# =========================
# 3) Vectorized transitions
# =========================

pf_df_step0 = pf_df_entry[["user_access_token", "entry_method"]].rename(
    columns={"entry_method": "pid_label"}
).copy()
pf_df_step0["step"] = 0

pf_df_all_steps = pd.concat(
    [pf_df_step0[["user_access_token", "step", "pid_label"]], pf_df_after[["user_access_token", "step", "pid_label"]]],
    ignore_index=True,
)
pf_df_all_steps.sort_values(["user_access_token", "step"], inplace=True)
pf_df_all_steps.reset_index(drop=True, inplace=True)

pf_df_all_steps["next_pid"] = pf_df_all_steps.groupby("user_access_token")["pid_label"].shift(-1)
pf_df_all_steps["next_step"] = pf_df_all_steps.groupby("user_access_token")["step"].shift(-1)

pf_df_trans = pf_df_all_steps.dropna(subset=["next_pid", "next_step"]).copy()
pf_df_trans = pf_df_trans.loc[pf_df_trans["next_step"] == pf_df_trans["step"] + 1].copy()
pf_df_trans = pf_df_trans.rename(
    columns={
        "step": "from_step",
        "pid_label": "from",
        "next_step": "to_step",
        "next_pid": "to",
    }
)
pf_df_trans["to_step"] = pf_df_trans["to_step"].astype("int16")

del pf_df_all_steps, pf_df_after, pf_df_step0, pf_df
gc.collect()


# =========================
# 4) Top-N per step + remap Other
# =========================

pf_top_per_step = {}
for pf_step in range(1, PF_MAX_STEPS + 1):
    pf_sl = pf_df_trans.loc[pf_df_trans["to_step"] == pf_step]
    if pf_sl.empty:
        pf_top_per_step[pf_step] = set(pf_entry_methods_sorted)
        continue
    pf_top_set = set(pf_sl["to"].value_counts().head(PF_TOP_METHODS).index)
    for pf_em in pf_entry_methods_sorted:
        pf_top_set.add(pf_em)
    pf_top_per_step[pf_step] = pf_top_set


def pf_remap(pf_lbl, pf_stp):
    if pf_stp == 0:
        return pf_lbl
    return pf_lbl if pf_lbl in pf_top_per_step.get(pf_stp, set()) else "Other"


pf_df_trans["from_r"] = pf_df_trans.apply(lambda r: pf_remap(r["from"], int(r["from_step"])), axis=1)
pf_df_trans["to_r"] = pf_df_trans.apply(lambda r: pf_remap(r["to"], int(r["to_step"])), axis=1)

pf_agg = pf_df_trans.groupby(["from_step", "to_step", "from_r", "to_r"]).size().reset_index(name="value")
pf_agg.columns = ["from_step", "to_step", "from", "to", "value"]
del pf_df_trans
gc.collect()


# =========================
# 5) Sankey nodes & links
# =========================

PF_PID_COLORS = pf_build_colors(
    set(pf_agg["from"]).union(pf_agg["to"]).union(pf_entry_methods_sorted)
)

pf_node_lookup, pf_all_nodes, pf_nid = {}, [], 0

for pf_s in range(PF_MAX_STEPS + 1):
    if pf_s == 0:
        pf_pids = [m for m in pf_entry_methods_sorted if m in pf_df_entry["entry_method"].values]
    else:
        pf_sl = pf_agg.loc[pf_agg["to_step"] == pf_s]
        if pf_sl.empty:
            continue
        pf_vol = pf_sl.groupby("to")["value"].sum().sort_values(ascending=False)
        pf_has_other = "Other" in pf_vol.index
        pf_entry_in_vol = [m for m in pf_entry_methods_sorted if m in pf_vol.index]
        pf_middle = [p for p in pf_vol.index if p not in pf_entry_methods_sorted and p != "Other"]
        pf_pids = pf_entry_in_vol + pf_middle + (["Other"] if pf_has_other else [])

    for pf_pid in pf_pids:
        pf_node_lookup[(pf_s, pf_pid)] = pf_nid
        pf_all_nodes.append({"id": pf_nid, "name": pf_pid, "step": pf_s})
        pf_nid += 1

pf_links_data = []
for _, pf_row in pf_agg.iterrows():
    pf_src = pf_node_lookup.get((int(pf_row["from_step"]), pf_row["from"]))
    pf_tgt = pf_node_lookup.get((int(pf_row["to_step"]), pf_row["to"]))
    if pf_src is not None and pf_tgt is not None:
        pf_links_data.append({"source": pf_src, "target": pf_tgt, "value": int(pf_row["value"])})

pf_step_totals = {0: pf_total_users}
for pf_s in range(1, PF_MAX_STEPS + 1):
    pf_v = int(pf_agg.loc[pf_agg["to_step"] == pf_s]["value"].sum())
    if pf_v > 0:
        pf_step_totals[pf_s] = pf_v

pf_method_per_step = {}
for pf_node_item in pf_all_nodes:
    pf_mn = pf_node_item["name"]
    pf_msv = pf_node_item["step"]
    if pf_msv == 0:
        pf_mu = int(pf_entry_counts.get(pf_mn, 0))
    else:
        pf_sl_tmp = pf_agg.loc[(pf_agg["to_step"] == pf_msv) & (pf_agg["to"] == pf_mn)]
        pf_mu = int(pf_sl_tmp["value"].sum()) if not pf_sl_tmp.empty else 0
    if pf_mn not in pf_method_per_step:
        pf_method_per_step[pf_mn] = {}
    pf_method_per_step[pf_mn][str(pf_msv)] = pf_mu

pf_mps_json = json.dumps(pf_method_per_step)
pf_entry_users_json = json.dumps({m: int(pf_entry_counts.get(m, 0)) for m in pf_entry_methods_sorted})


# =========================
# 6) KPI cards
# =========================

def pf_make_kpi(pf_label, pf_tokens, pf_step_total, pf_accent, pf_is_best=False):
    pf_pct = pf_tokens / pf_step_total * 100 if pf_step_total > 0 else 0
    pf_badge = ' <span class="pf-kpi-badge-inline">BEST</span>' if pf_is_best else ""
    return (
        f'<div class="pf-kpi-card" style="--accent:{pf_accent}">'
        f'<div class="pf-kpi-label">{pf_label} <span class="pf-kpi-step">step share</span></div>'
        f'<div class="pf-kpi-value-row"><span class="pf-kpi-value">{pf_pct:.1f}%</span>{pf_badge}</div>'
        f'<div class="pf-kpi-note">{pf_tokens:,} of {pf_step_total:,} step tokens</div>'
        f"</div>"
    )


pf_step_for_kpi = 1
pf_step_total_kpi = int(pf_step_totals.get(pf_step_for_kpi, 0))

pf_step_method_counts = (
    pf_agg.loc[pf_agg["to_step"] == pf_step_for_kpi]
    .groupby("to", as_index=False)["value"]
    .sum()
    .sort_values("value", ascending=False)
)

pf_kpi_top = pf_step_method_counts.head(5).copy()

pf_kpi_data = []
for _, _pf_row in pf_kpi_top.iterrows():
    _pf_method = str(_pf_row["to"])
    _pf_tokens = int(_pf_row["value"])
    _pf_pct = (_pf_tokens / pf_step_total_kpi * 100) if pf_step_total_kpi > 0 else 0
    pf_kpi_data.append((_pf_method, _pf_tokens, _pf_pct))

pf_best_method = max(pf_kpi_data, key=lambda x: x[2])[0] if pf_kpi_data else ""

pf_kpi_html = ""
for pf_m, pf_tokens, _ in pf_kpi_data:
    pf_kpi_html += pf_make_kpi(
        pf_m,
        pf_tokens,
        pf_step_total_kpi,
        PF_PID_COLORS.get(pf_m, XS.C_OTHER),
        pf_is_best=(pf_m == pf_best_method),
    )

# =========================
# 7) Step labels
# =========================

pf_ord = ["1st", "2nd", "3rd", "4th", "5th", "6th", "7th", "8th", "9th", "10th", "11th"]
pf_step_labels = [f"{pf_ord[i]} step" for i in range(PF_MAX_STEPS + 1)]
pf_step_sub = [f"{pf_total_users:,} sessions"] + [
    f"{pf_step_totals.get(pf_s, 0):,} sessions" for pf_s in range(1, PF_MAX_STEPS + 1)
]

pf_subtitle = (
    f"My.Games merchant 799928 · US · {pf_date_range}"
)


# =========================
# 8) Build HTML (Xsolla-themed CSS, d3 Sankey unchanged)
# =========================

_bg = XS.BG_PAGE
_card = XS.BG_PLOT
_primary = XS.TEXT_PRIMARY
_secondary = XS.TEXT_SECONDARY
_border = XS.BORDER
_border_light = XS.BORDER_LIGHT

pf_html = """<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<link href="https://fonts.googleapis.com/css2?family=Open+Sans:wght@400;600;700&family=Roboto:wght@400;500;700&family=Roboto+Mono:wght@400;500&display=swap" rel="stylesheet">
<script src="https://cdnjs.cloudflare.com/ajax/libs/d3/7.8.5/d3.min.js"></script>
<script src="https://cdnjs.cloudflare.com/ajax/libs/d3-sankey/0.12.3/d3-sankey.min.js"></script>
<style>
  *,*::before,*::after{box-sizing:border-box;margin:0;padding:0}
  body{background:__BG__;font-family:__FONT_SANS__;color:__PRIMARY__;padding:24px 16px 36px 32px}
  .pfw{max-width:1700px;margin:0 auto}
  h1{font-size:19px;font-weight:700;color:__PRIMARY__;letter-spacing:-.4px;line-height:1.2;font-family:'Open Sans',sans-serif}
  .pf-sub{font-size:11px;color:__SECONDARY__;margin-top:3px;margin-bottom:16px}

  .pf-kr{display:flex;gap:8px;flex-wrap:wrap;margin-bottom:20px}
  .pf-kpi-card{background:__CARD__;border:1px solid __BORDER__;border-radius:10px;padding:11px 14px;
    min-width:110px;flex:1;position:relative;overflow:hidden;box-shadow:0 1px 4px rgba(0,0,0,.06)}
  .pf-kpi-card::before{content:'';position:absolute;top:0;left:0;right:0;height:2px;background:var(--accent,__ACCENT__)}
  .pf-kpi-label{font-size:10px;color:__SECONDARY__;font-weight:500;line-height:1.3}
  .pf-kpi-step{display:inline-block;background:__SUBTLE__;color:__SECONDARY__;font-size:8px;font-weight:600;
    padding:1px 5px;border-radius:4px;letter-spacing:.05em;margin-left:3px;vertical-align:middle}
  .pf-kpi-value-row{display:flex;align-items:baseline;gap:7px;margin-top:2px}
  .pf-kpi-value{font-size:22px;font-weight:700;color:__PRIMARY__;font-family:'Roboto Mono',monospace;line-height:1.2}
  .pf-kpi-badge-inline{display:inline-block;background:rgba(22,163,74,.12);color:__GOOD__;font-size:9px;font-weight:700;
    padding:2px 7px;border-radius:99px;letter-spacing:.06em;border:1px solid rgba(22,163,74,.35);
    white-space:nowrap;vertical-align:middle}
  .pf-kpi-note{font-size:9px;color:__MUTED__;margin-top:2px}

  .pf-cw{background:__CARD__;border:1px solid __BORDER__;border-radius:14px;padding:20px 16px;box-shadow:0 1px 4px rgba(0,0,0,.06)}
  .pf-cw svg{display:block;width:100%}

  #pf-tt{position:fixed;background:__CARD__;border:1px solid __BORDER__;border-radius:10px;padding:10px 14px;
    font-size:12px;pointer-events:none;opacity:0;transition:opacity .15s;
    box-shadow:0 8px 32px rgba(0,0,0,.12);z-index:1000;max-width:340px;line-height:1.5;font-family:__FONT_SANS2__}
  .pf-ttt{font-weight:600;color:__PRIMARY__;margin-bottom:6px;font-size:12px}
  .pf-ttr{display:flex;justify-content:space-between;gap:16px;color:__SECONDARY__;font-size:11px;padding:1px 0}
  .pf-ttr span:last-child{color:__PRIMARY__;font-family:'Roboto Mono',monospace;font-weight:500}
  .pf-tts{color:__SECONDARY__;margin:7px 0 3px;font-size:10px;border-top:1px solid __BORDER_LIGHT__;padding-top:6px;
    font-weight:600;letter-spacing:.04em;text-transform:uppercase}

  .pf-lr{display:flex;flex-wrap:wrap;gap:10px;margin-top:14px;padding-top:12px;border-top:1px solid __BORDER_LIGHT__}
  .pf-li{display:flex;align-items:center;gap:5px;font-size:11px;color:__SECONDARY__}
  .pf-ld{width:8px;height:8px;border-radius:3px;flex-shrink:0}

  .sankey-link{cursor:pointer}.sankey-link path{transition:opacity .2s}
  .sankey-node{cursor:pointer}.pf-col-h{pointer-events:none}

  .pf-metrics-desc{margin-top:24px;padding-top:20px;border-top:1px solid __BORDER_LIGHT__}
  .pf-md-title{font-size:10px;font-weight:600;color:__SECONDARY__;text-transform:uppercase;
    letter-spacing:.08em;margin-bottom:14px;font-family:'Open Sans',sans-serif}
  .pf-md-grid{display:grid;grid-template-columns:repeat(3,1fr);gap:16px}
  .pf-md-item{background:__SUBTLE__;border:1px solid __BORDER_LIGHT__;border-radius:8px;padding:12px 14px}
  .pf-md-name{font-size:11px;font-weight:700;color:__ACCENT__;margin-bottom:6px;
    font-family:'Open Sans',sans-serif;letter-spacing:-.01em}
  .pf-md-text{font-size:11px;color:__SECONDARY__;line-height:1.65;font-family:'Open Sans',sans-serif}
  .pf-md-text em{font-style:normal;color:__PRIMARY__;font-weight:600}
</style>
</head>
<body>
<div class="pfw">
  <h1>Payment Flow • My.Games</h1>
  <div class="pf-sub">__SUBTITLE__</div>
  <div class="pf-kr">__KPI_HTML__</div>
  <div class="pf-cw">
    <div id="pf-chart"></div>
    <div class="pf-lr" id="pf-legend"></div>
    <div class="pf-metrics-desc">
      <div class="pf-md-title">Metrics glossary · hover on any flow or node to see values</div>
      <div class="pf-md-grid">
        <div class="pf-md-item">
          <div class="pf-md-name">Preference Rate prev step %</div>
          <div class="pf-md-text">
            Share of sessions on a transition among <em>all sessions leaving the same method node</em> at the previous step.
          </div>
        </div>
        <div class="pf-md-item">
          <div class="pf-md-name">Total Preference Rate %</div>
          <div class="pf-md-text">
            Share on a transition out of <em>all entry sessions</em> for that method at step 0.
          </div>
        </div>
        <div class="pf-md-item">
          <div class="pf-md-name">% of step total</div>
          <div class="pf-md-text">
            Share among <em>all sessions present at that step number</em>, across methods.
          </div>
        </div>
      </div>
    </div>
  </div>
</div>
<div id="pf-tt"></div>

<script>
const pf_ND=__NODES__,pf_LK=__LINKS__,pf_ST=__STEP_TOT__,pf_SL=__STEP_LBL__,
      pf_SS=__STEP_SUB__,pf_CL=__COLORS__,pf_MS=__MAX_STEPS__,
      pf_EU=__ENTRY_USERS__,pf_MPS=__METHOD_PER_STEP__;

const pf_W=Math.max(980,(window.innerWidth||1500)-40),pf_H=532,
      pf_M={top:72,right:130,bottom:24,left:130},
      pf_CW=pf_W-pf_M.left-pf_M.right,pf_CH=pf_H-pf_M.top-pf_M.bottom,
      pf_NW=14,pf_NP=8,pf_MH=4;

const pf_svg=d3.select('#pf-chart').append('svg').attr('viewBox',`0 0 ${pf_W} ${pf_H}`);
const pf_defs=pf_svg.append('defs');
const pf_g=pf_svg.append('g').attr('transform',`translate(${pf_M.left},${pf_M.top})`);

const pf_sk=d3.sankey().nodeId(d=>d.id).nodeAlign(d=>d.step)
  .nodeSort((a,b)=>{
    if(a.step!==b.step)return 0;
    if(a.step===0){
      const ua=pf_EU[a.name]||0,ub=pf_EU[b.name]||0;
      return ub-ua;
    }
    if(a.name==='Other'&&b.name!=='Other')return 1;
    if(b.name==='Other'&&a.name!=='Other')return-1;
    const va=(a.targetLinks||[]).reduce((s,l)=>s+l.value,0)||(a.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
    const vb=(b.targetLinks||[]).reduce((s,l)=>s+l.value,0)||(b.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
    return vb-va;
  }).nodeWidth(pf_NW).nodePadding(pf_NP).iterations(64).extent([[0,0],[pf_CW,pf_CH]]);

const pf_nodes=pf_ND.map(d=>({...d})),pf_links=pf_LK.map(d=>({...d}));
const pf_gr=pf_sk({nodes:pf_nodes,links:pf_links});

// Force exact X position per step so node columns stay within step boundaries.
const pf_stepMax=Math.max(1,pf_MS);
pf_gr.nodes.forEach(n=>{
  const x=(n.step/pf_stepMax)*pf_CW;
  n.x0=x;
  n.x1=x+pf_NW;
});
pf_sk.update(pf_gr);

pf_gr.links.forEach((lk,i)=>{
  const g2=pf_defs.append('linearGradient').attr('id',`pfg${i}`).attr('gradientUnits','userSpaceOnUse')
    .attr('x1',lk.source.x1).attr('x2',lk.target.x0);
  g2.append('stop').attr('offset','0%').attr('stop-color',pf_CL[lk.source.name]||'#94a3b8');
  g2.append('stop').attr('offset','100%').attr('stop-color',pf_CL[lk.target.name]||'#94a3b8');
});

const pf_sx={};
pf_gr.nodes.forEach(n=>{if(!(n.step in pf_sx))pf_sx[n.step]=n.x0+pf_NW/2;});
for(let s=0;s<=pf_MS;s++){
  const cx=pf_sx[s];if(cx===undefined)continue;
  const hg=pf_g.append('g').attr('class','pf-col-h');
  hg.append('rect').attr('x',cx-42).attr('y',-pf_M.top+4).attr('width',84).attr('height',20).attr('rx',5)
    .attr('fill',s===0?'__S12__':'__SUBTLE__').attr('stroke',s===0?'__S7__':'__BORDER__').attr('stroke-width',1);
  hg.append('text').attr('x',cx).attr('y',-pf_M.top+17).attr('text-anchor','middle')
    .attr('fill',s===0?'__S5__':'__SECONDARY__').attr('font-size',10)
    .attr('font-family', "'Open Sans',sans-serif").attr('font-weight',600).attr('letter-spacing','.04em')
    .text((pf_SL[s]||'').toUpperCase());
  hg.append('text').attr('x',cx).attr('y',-pf_M.top+40).attr('text-anchor','middle')
    .attr('fill','__SECONDARY__').attr('font-size',10).attr('font-family', "'Roboto Mono',monospace")
    .text(pf_SS[s]||'');
  if(s>0)pf_g.append('line').attr('x1',cx).attr('y1',-pf_M.top+50).attr('x2',cx).attr('y2',pf_CH)
    .attr('stroke','__BORDER__').attr('stroke-width',1).attr('stroke-dasharray','3,5');
}

const pf_lp=d3.sankeyLinkHorizontal();
const pf_ls=pf_g.append('g').selectAll('.sankey-link').data(pf_gr.links).enter().append('g').attr('class','sankey-link');
pf_ls.append('path').attr('d',pf_lp).attr('fill','none')
  .attr('stroke',(d,i)=>`url(#pfg${i})`).attr('stroke-width',d=>Math.max(1,d.width)).attr('opacity',.28);

const pf_ns=pf_g.append('g').selectAll('.sankey-node').data(pf_gr.nodes).enter().append('g').attr('class','sankey-node')
  .attr('transform',d=>`translate(${d.x0},${d.y0})`);
pf_ns.append('rect').attr('width',pf_NW).attr('height',d=>Math.max(pf_MH,d.y1-d.y0))
  .attr('rx',3).attr('fill',d=>pf_CL[d.name]||'#94a3b8').attr('opacity',.88);
pf_ns.filter(d=>d.step===0||(d.name in pf_EU&&pf_EU[d.name]>0)).append('rect')
  .attr('width',pf_NW).attr('height',d=>Math.max(pf_MH,d.y1-d.y0))
  .attr('rx',3).attr('fill','none')
  .attr('stroke',d=>pf_CL[d.name]||'#94a3b8').attr('stroke-width',1.2).attr('opacity',.5);

const pf_lg=pf_g.append('g');
pf_gr.nodes.forEach(d=>{
  const h=Math.max(pf_MH,d.y1-d.y0),cy=d.y0+h/2;
  const isEntry = d.step===0;
  if(h<6&&!isEntry)return;
  const xpos = isEntry ? d.x0-7 : d.x1+7;
  const an   = isEntry ? 'end'  : 'start';
  const ll   = pf_lg.append('g');
  ll.append('text').attr('x',xpos).attr('y',cy-(h>14?5:0)).attr('text-anchor',an)
    .attr('fill',pf_CL[d.name]||'__SECONDARY__')
    .attr('font-size',isEntry?12:10).attr('font-weight',isEntry?600:400)
    .attr('font-family', "'Open Sans',sans-serif").attr('dominant-baseline',h<=14?'middle':'auto')
    .text(d.name);
  if(isEntry && h>14){
    ll.append('text').attr('x',xpos).attr('y',cy+6).attr('text-anchor',an)
      .attr('fill',pf_CL[d.name]||'__SECONDARY__').attr('font-size',9)
      .attr('font-family', "'Roboto Mono',monospace").attr('font-weight',500)
      .text(`${(pf_EU[d.name]||0).toLocaleString()} entry`);
    return;
  }
  if(h>14){
    const prevU = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step-1)];
    const thisU = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step)]||0;
    if(prevU&&prevU>0){
      const dropPct=((prevU-thisU)/prevU*100).toFixed(1);
      ll.append('text').attr('x',xpos).attr('y',cy+6).attr('text-anchor',an)
        .attr('fill','__BAD__').attr('font-size',8.5)
        .attr('font-family', "'Roboto Mono',monospace").attr('font-weight',500)
        .text(`drop \\u2193 ${dropPct}%`);
    }
  }
});

const pf_tt=d3.select('#pf-tt'),pf_fmt=n=>n.toLocaleString();
const pf_sT=(h,e)=>pf_tt.html(h).style('opacity',1).style('left',(e.clientX+14)+'px').style('top',(e.clientY-10)+'px');
const pf_hT=()=>pf_tt.style('opacity',0);
const pf_mT=e=>pf_tt.style('left',(e.clientX+14)+'px').style('top',(e.clientY-10)+'px');

pf_ls.on('mouseover',function(ev,d){
  pf_ls.selectAll('path').attr('opacity',.05);d3.select(this).select('path').attr('opacity',.75);
  const srcTot   = (d.source.sourceLinks||[]).reduce((s,l)=>s+l.value,0)||(pf_EU[d.source.name]||1);
  const prevPct  = (d.value/srcTot*100).toFixed(1);
  const entryTot = pf_EU[d.source.name]||srcTot;
  const totalPct = (d.value/entryTot*100).toFixed(1);
  const stepTot  = d.source.step===0?(pf_EU[d.source.name]||1):(pf_ST[String(d.source.step)]||1);
  const stepPct  = (d.value/stepTot*100).toFixed(1);
  pf_sT(`
    <div class="pf-ttt">${d.source.name} \\u2192 ${d.target.name}</div>
    <div class="pf-ttr"><span>Departed \\u2192 Arriving</span><span>${pf_fmt(d.value)}</span></div>
    <div class="pf-ttr"><span>Preference Rate prev step %</span><span>${prevPct}%</span></div>
    <div class="pf-ttr"><span>Total Preference Rate %</span><span>${totalPct}%</span></div>
    <div class="pf-ttr"><span>% of step total</span><span>${stepPct}%</span></div>
    <div class="pf-tts">\\u2191 SOURCE (step ${d.source.step+1})</div>
    <div class="pf-ttr"><span>${d.source.name} sessions</span><span>${pf_fmt(srcTot)}</span></div>
  `,ev);
}).on('mousemove',function(ev){pf_mT(ev);}).on('mouseout',function(){pf_ls.selectAll('path').attr('opacity',.28);pf_hT();});

pf_ns.on('mouseover',function(ev,d){
  const cL=new Set(),cN=new Set([d.id]);
  (d.sourceLinks||[]).forEach(l=>{cL.add(l.index);cN.add(l.target.id);});
  (d.targetLinks||[]).forEach(l=>{cL.add(l.index);cN.add(l.source.id);});
  pf_ls.selectAll('path').attr('opacity',(ld,i)=>cL.has(i)?.68:.04);
  pf_ns.selectAll('rect:first-child').attr('opacity',nd=>cN.has(nd.id)?.95:.18);
  const inc     = (d.targetLinks||[]).reduce((s,l)=>s+l.value,0);
  const out     = (d.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
  const stepTot = d.step===0?(pf_EU[d.name]||1):(pf_ST[String(d.step)]||1);
  const prevSame     = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step-1)];
  const thisSame     = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step)]||inc;
  const nextSameLink = (d.sourceLinks||[]).find(l=>l.target.name===d.name);
  const nextSame     = nextSameLink?nextSameLink.value:0;
  const prefPrevPct  = prevSame&&prevSame>0?(thisSame/prevSame*100).toFixed(1):null;
  const sortedOut = [...(d.sourceLinks||[])].sort((a,b)=>b.value-a.value).slice(0,5);
  const nextHtml  = out>0?sortedOut.map(l=>
    `<div class="pf-ttr"><span>\\u2192 ${l.target.name}</span><span>${pf_fmt(l.value)} (${(l.value/out*100).toFixed(1)}%)</span></div>`
  ).join(''):'';
  const tgtInc  = (d.targetLinks||[]).reduce((s,l)=>s+l.value,0);
  const sortedSrc = [...(d.targetLinks||[])].sort((a,b)=>b.value-a.value).slice(0,3);
  const srcHtml = tgtInc>0?sortedSrc.map(l=>
    `<div class="pf-ttr"><span>\\u2190 ${l.source.name}</span><span>${pf_fmt(l.value)} (${(l.value/tgtInc*100).toFixed(1)}%)</span></div>`
  ).join(''):'';
  if(d.step===0){
    pf_sT(`
      <div class="pf-ttt">${d.name} \\u00b7 Step 1</div>
      <div class="pf-ttr"><span>Entry sessions</span><span>${pf_fmt(pf_EU[d.name]||0)}</span></div>
      <div class="pf-ttr"><span>With next step</span><span>${pf_fmt(out)}</span></div>
      ${nextHtml?`<div class="pf-tts">\\u2192 Next</div>${nextHtml}`:''}
    `,ev);
    return;
  }
  pf_sT(`
    <div class="pf-ttt">${d.name} \\u00b7 Step ${d.step+1}</div>
    ${prevSame?`<div class="pf-ttr"><span>${d.name} at prev step</span><span>${pf_fmt(prevSame)}</span></div>`:''}
    <div class="pf-ttr"><span>${d.name} at this step</span><span>${pf_fmt(thisSame)}</span></div>
    ${prefPrevPct?`<div class="pf-ttr"><span>Preference Rate prev step %</span><span>${prefPrevPct}%</span></div>`:''}
    ${nextSame>0?`<div class="pf-ttr"><span>${d.name} at next step</span><span>${pf_fmt(nextSame)}</span></div>`:''}
    <div class="pf-ttr"><span>Share of step</span><span>${(Math.max(inc,out)/stepTot*100).toFixed(1)}%</span></div>
    ${nextHtml?`<div class="pf-tts">\\u2192 Next</div>${nextHtml}`:''}
    ${srcHtml?`<div class="pf-tts">\\u2190 Top sources</div>${srcHtml}`:''}
  `,ev);
}).on('mousemove',function(ev){pf_mT(ev);})
  .on('mouseout',function(){pf_ls.selectAll('path').attr('opacity',.28);pf_ns.selectAll('rect:first-child').attr('opacity',.88);pf_hT();});

const pf_le=document.getElementById('pf-legend'),pf_un=new Set(pf_gr.nodes.map(n=>n.name));
Object.entries(pf_CL).filter(([n])=>pf_un.has(n)).forEach(([n,c])=>{
  const it=document.createElement('div');it.className='pf-li';
  it.innerHTML=`<div class="pf-ld" style="background:${c}"></div>${n}`;pf_le.appendChild(it);
});

pf_ls.selectAll('path')
  .attr('stroke-dasharray',function(){return this.getTotalLength()+' '+this.getTotalLength();})
  .attr('stroke-dashoffset',function(){return this.getTotalLength();})
  .transition().duration(1400).delay((d,i)=>50+i*4).ease(d3.easeCubicOut)
  .attr('stroke-dashoffset',0).on('end',function(){d3.select(this).attr('stroke-dasharray',null);});
pf_ns.attr('opacity',0).transition().duration(400).delay(d=>d.step*70+120).attr('opacity',1);
</script></body></html>"""

pf_html = (
    pf_html.replace("__BG__", _bg)
    .replace("__CARD__", _card)
    .replace("__PRIMARY__", _primary)
    .replace("__SECONDARY__", _secondary)
    .replace("__BORDER__", _border)
    .replace("__BORDER_LIGHT__", _border_light)
    .replace("__SUBTLE__", XS.BG_SUBTLE)
    .replace("__S12__", XS.S12)
    .replace("__S7__", XS.S7)
    .replace("__S5__", XS.S5)
    .replace("__ACCENT__", XS.S7)
    .replace("__MUTED__", XS.TEXT_HINT)
    .replace("__GOOD__", XS.GOOD)
    .replace("__BAD__", XS.BAD)
    .replace("__FONT_SANS__", "'Open Sans', 'Roboto', sans-serif")
    .replace("__FONT_SANS2__", "'Open Sans', 'Roboto', sans-serif")
    .replace("__SUBTITLE__", pf_subtitle)
    .replace("__KPI_HTML__", pf_kpi_html)
    .replace("__NODES__", json.dumps(pf_all_nodes))
    .replace("__LINKS__", json.dumps(pf_links_data))
    .replace("__STEP_TOT__", json.dumps({str(k): v for k, v in pf_step_totals.items()}))
    .replace("__STEP_LBL__", json.dumps(pf_step_labels))
    .replace("__STEP_SUB__", json.dumps(pf_step_sub))
    .replace("__COLORS__", json.dumps(PF_PID_COLORS))
    .replace("__MAX_STEPS__", str(PF_MAX_STEPS))
    .replace("__ENTRY_USERS__", pf_entry_users_json)
    .replace("__METHOD_PER_STEP__", pf_mps_json)
)


# =========================
# 9) Confluence export
# =========================

pf_conf = (
    pf_html.replace("background:" + _bg + ";", "background:transparent;")
    .replace("padding:24px 16px 36px 32px;", "padding:8px 4px 16px;")
    .replace("max-width:1700px;", "max-width:100%;")
)

pf_export = (
    '<div style="font-family:sans-serif;margin-top:20px;border-radius:12px;'
    'border:1px solid #e2e8f0;overflow:hidden;box-shadow:0 1px 6px rgba(0,0,0,.07);">'
    '<button onclick="var b=document.getElementById(\'pfcb\');var a=document.getElementById(\'pfca\');'
    'var o=b.style.display!==\'none\';b.style.display=o?\'none\':\'block\';'
    'a.style.transform=o?\'rotate(0deg)\':\'rotate(180deg)\';"'
    ' style="width:100%;background:#f8fafc;border:none;border-bottom:1px solid #e2e8f0;'
    'padding:14px 20px;display:flex;align-items:center;justify-content:space-between;cursor:pointer;">'
    '<div><div style="font-size:13px;font-weight:600;color:#0f172a;">Confluence HTML'
    '<span style="background:#EBFCFF;color:#1B8398;font-size:10px;font-weight:600;'
    'padding:2px 8px;border-radius:99px;margin-left:6px;">READY TO PASTE</span></div>'
    '<div style="font-size:11px;color:#94a3b8;margin-top:1px;">'
    'Copy and paste into HTML Macro (height: 1000px)</div></div>'
    '<span id="pfca" style="transition:transform .25s;font-size:18px;color:#94a3b8;">&#8964;</span>'
    '</button>'
    '<div id="pfcb" style="display:none;background:#fff;">'
    '<div style="position:relative;background:#f8fafc;border-bottom:1px solid #f1f5f9;">'
    '<textarea id="pfcc" readonly style="width:100%;height:140px;font-family:monospace;font-size:11px;'
    'color:#475569;background:transparent;border:none;padding:14px 20px;'
    'resize:none;outline:none;line-height:1.6;box-sizing:border-box;">'
    + pf_conf.replace("'", "&#39;")
    + '</textarea>'
    + '<div id="pfck" style="position:absolute;bottom:8px;right:14px;font-size:10px;'
    'color:#cbd5e1;font-family:monospace;pointer-events:none;"></div></div>'
    '<div style="padding:12px 20px;display:flex;align-items:center;gap:12px;">'
    '<button id="pfcpb" onclick="pfCopy()" style="background:#31CAE8;color:#fff;border:none;'
    'border-radius:8px;padding:9px 20px;font-size:13px;font-weight:600;cursor:pointer;transition:background .2s;">'
    'Copy to clipboard</button>'
    '<span id="pfcps" style="font-size:12px;color:#94a3b8;"></span>'
    '</div></div></div>'
    '<script>'
    'var pfta=document.getElementById("pfcc");'
    'document.getElementById("pfck").textContent=(pfta.value.length/1024).toFixed(1)+" KB";'
    'function pfCopy(){'
    'navigator.clipboard.writeText(pfta.value).then(function(){'
    'var btn=document.getElementById("pfcpb"),st=document.getElementById("pfcps");'
    'btn.style.background="#16a34a";btn.textContent="Copied!";'
    'st.textContent="Copied";st.style.color="#16a34a";'
    'setTimeout(function(){btn.style.background="#31CAE8";btn.textContent="Copy to clipboard";st.textContent="";},3000);'
    '}).catch(function(){pfta.select();document.execCommand("copy");});}'
    '</script>'
)

# Jupyter sanitizes inline <script> in HTML(); serve the same file over HTTP and embed via IFrame (D3 works).
import functools
import http.server
import socketserver
import threading
import webbrowser

_pf_preview_path = (Path.cwd() / "payment_flow_sankey_preview.html").resolve()
_pf_preview_path.write_text(pf_html, encoding="utf-8")

class _PfTCPServer(socketserver.TCPServer):
    allow_reuse_address = True


class _PfQuietHandler(http.server.SimpleHTTPRequestHandler):
    def log_message(self, format, *args):
        return


_pf_root = _pf_preview_path.parent
_pf_httpd = None
_pf_port = None
for _pf_try_port in range(8765, 8795):
    try:
        _pf_handler = functools.partial(
            _PfQuietHandler,
            directory=str(_pf_root),
        )
        _pf_httpd = _PfTCPServer(("127.0.0.1", _pf_try_port), _pf_handler)
        _pf_port = _pf_try_port
        break
    except OSError:
        continue

if _pf_httpd is not None and _pf_port is not None:
    threading.Thread(target=_pf_httpd.serve_forever, daemon=True).start()
    _pf_iframe_url = f"http://127.0.0.1:{_pf_port}/{_pf_preview_path.name}"
    print("Embedded chart (HTTP iframe, scripts allowed):", _pf_iframe_url)
    display(IFrame(src=_pf_iframe_url, width="100%", height=620))
else:
    print("Could not start local HTTP preview; open the HTML file or use the VS Code task.")














```

```python
# Payment Flow (df_payment_flow_my_games) — Template 1 structure, Xsolla palette, session step logic

import gc
import json
import sys
from pathlib import Path

import pandas as pd
from IPython.display import HTML, IFrame, display

_cwd = Path.cwd()
_xsolla_root = _cwd if (_cwd / "utils" / "style.py").is_file() else _cwd.parent
if str(_xsolla_root) not in sys.path:
    sys.path.insert(0, str(_xsolla_root))
from utils.style import XS

# =========================
# 1) Parameters
# =========================

PF_MAX_STEPS = 10
PF_TOP_METHODS = 10
PF_DATE_COL = "internal_date"

_xs_pal = [XS.S3, XS.S5, XS.S7, XS.S9, XS.S6, XS.S8, XS.S4, XS.S10]


def pf_pid_label(raw):
    if pd.isna(raw):
        return "Unknown"
    s = str(raw).strip()
    parts = s.split(".", 1)
    return parts[1].strip() if len(parts) > 1 else s


def pf_row_flow_label(row):
    if int(row["sankey_step"]) == 0:
        return pf_pid_label(row["pid"])
    return pf_pid_label(row["bill_pid"])


def pf_build_colors(names):
    out = {"Other": XS.C_OTHER}
    names = sorted(n for n in names if n and n != "Other")
    for i, n in enumerate(names):
        out[n] = _xs_pal[i % len(_xs_pal)]
    return out


# =========================
# 2) Data preparation
# =========================

pf_df = df_payment_flow_my_games.copy()
pf_df = pf_df.dropna(subset=["pid"]).copy()
pf_df["internal_time"] = pd.to_datetime(pf_df["internal_time"], errors="coerce")
pf_df = pf_df.sort_values(["user_access_token", "steps"]).reset_index(drop=True)
pf_df["sankey_step"] = pf_df.groupby("user_access_token", sort=False).cumcount().astype("int64")
pf_df["flow_label"] = pf_df.apply(pf_row_flow_label, axis=1)
pf_df["pid_label"] = pf_df["flow_label"]
gc.collect()

try:
    pf_date_range = f"{pf_df[PF_DATE_COL].min()} – {pf_df[PF_DATE_COL].max()}"
except Exception:
    pf_date_range = ""

pf_df_entry = (
    pf_df.loc[pf_df["sankey_step"] == 0, ["user_access_token", "pid_label"]]
    .drop_duplicates(subset=["user_access_token"])
    .rename(columns={"pid_label": "entry_method"})
    .copy()
)

# Entry step labels: keep top-5 methods, group all others into Other.
pf_entry_counts_all = pf_df_entry["entry_method"].value_counts()
pf_entry_top5 = set(pf_entry_counts_all.head(5).index)
pf_entry_mandatory = {"Bank card", "Apple Pay", "PayPal", "Google Pay"}
pf_entry_allowed = (pf_entry_top5 | pf_entry_mandatory) - {"Unknown"}
pf_df_entry["entry_method"] = pf_df_entry["entry_method"].apply(
    lambda x: x if x in pf_entry_allowed else "Other"
)

pf_entry_counts = pf_df_entry["entry_method"].value_counts()
pf_entry_methods_sorted = pf_entry_counts.index.tolist()
pf_total_users = len(pf_df_entry)

pf_df_after = pf_df.loc[(pf_df["sankey_step"] >= 1) & (pf_df["done_just"] == 1) & (pf_df["bill_pid"].notna())].copy()
pf_df_after = pf_df_after.merge(
    pf_df_entry[["user_access_token", "entry_method"]], on="user_access_token", how="inner"
)
pf_df_after.sort_values(["user_access_token", "sankey_step"], inplace=True)
pf_df_after.reset_index(drop=True, inplace=True)
pf_df_after["step"] = pf_df_after["sankey_step"].astype("int16")
pf_df_after = pf_df_after.loc[pf_df_after["step"] <= PF_MAX_STEPS].copy()
gc.collect()


# =========================
# 3) Vectorized transitions
# =========================

pf_df_step0 = pf_df_entry[["user_access_token", "entry_method"]].rename(
    columns={"entry_method": "pid_label"}
).copy()
pf_df_step0["step"] = 0

pf_df_all_steps = pd.concat(
    [pf_df_step0[["user_access_token", "step", "pid_label"]], pf_df_after[["user_access_token", "step", "pid_label"]]],
    ignore_index=True,
)
pf_df_all_steps.sort_values(["user_access_token", "step"], inplace=True)
pf_df_all_steps.reset_index(drop=True, inplace=True)

pf_df_all_steps["next_pid"] = pf_df_all_steps.groupby("user_access_token")["pid_label"].shift(-1)
pf_df_all_steps["next_step"] = pf_df_all_steps.groupby("user_access_token")["step"].shift(-1)

pf_df_trans = pf_df_all_steps.dropna(subset=["next_pid", "next_step"]).copy()
pf_df_trans = pf_df_trans.loc[pf_df_trans["next_step"] == pf_df_trans["step"] + 1].copy()
pf_df_trans = pf_df_trans.rename(
    columns={
        "step": "from_step",
        "pid_label": "from",
        "next_step": "to_step",
        "next_pid": "to",
    }
)
pf_df_trans["to_step"] = pf_df_trans["to_step"].astype("int16")

del pf_df_all_steps, pf_df_after, pf_df_step0, pf_df
gc.collect()


# =========================
# 4) Top-N per step + remap Other
# =========================

pf_top_per_step = {}
for pf_step in range(1, PF_MAX_STEPS + 1):
    pf_sl = pf_df_trans.loc[pf_df_trans["to_step"] == pf_step]
    if pf_sl.empty:
        pf_top_per_step[pf_step] = set(pf_entry_methods_sorted)
        continue
    pf_top_set = set(pf_sl["to"].value_counts().head(PF_TOP_METHODS).index)
    for pf_em in pf_entry_methods_sorted:
        pf_top_set.add(pf_em)
    pf_top_per_step[pf_step] = pf_top_set


def pf_remap(pf_lbl, pf_stp):
    if pf_stp == 0:
        return pf_lbl
    return pf_lbl if pf_lbl in pf_top_per_step.get(pf_stp, set()) else "Other"


pf_df_trans["from_r"] = pf_df_trans.apply(lambda r: pf_remap(r["from"], int(r["from_step"])), axis=1)
pf_df_trans["to_r"] = pf_df_trans.apply(lambda r: pf_remap(r["to"], int(r["to_step"])), axis=1)

pf_agg = pf_df_trans.groupby(["from_step", "to_step", "from_r", "to_r"]).size().reset_index(name="value")
pf_agg.columns = ["from_step", "to_step", "from", "to", "value"]
del pf_df_trans
gc.collect()


# =========================
# 5) Sankey nodes & links
# =========================

PF_PID_COLORS = pf_build_colors(
    set(pf_agg["from"]).union(pf_agg["to"]).union(pf_entry_methods_sorted)
)

pf_node_lookup, pf_all_nodes, pf_nid = {}, [], 0

for pf_s in range(PF_MAX_STEPS + 1):
    if pf_s == 0:
        pf_pids = [m for m in pf_entry_methods_sorted if m in pf_df_entry["entry_method"].values]
    else:
        pf_sl = pf_agg.loc[pf_agg["to_step"] == pf_s]
        if pf_sl.empty:
            continue
        pf_vol = pf_sl.groupby("to")["value"].sum().sort_values(ascending=False)
        pf_has_other = "Other" in pf_vol.index
        pf_entry_in_vol = [m for m in pf_entry_methods_sorted if m in pf_vol.index]
        pf_middle = [p for p in pf_vol.index if p not in pf_entry_methods_sorted and p != "Other"]
        pf_pids = pf_entry_in_vol + pf_middle + (["Other"] if pf_has_other else [])

    for pf_pid in pf_pids:
        pf_node_lookup[(pf_s, pf_pid)] = pf_nid
        pf_all_nodes.append({"id": pf_nid, "name": pf_pid, "step": pf_s})
        pf_nid += 1

pf_links_data = []
for _, pf_row in pf_agg.iterrows():
    pf_src = pf_node_lookup.get((int(pf_row["from_step"]), pf_row["from"]))
    pf_tgt = pf_node_lookup.get((int(pf_row["to_step"]), pf_row["to"]))
    if pf_src is not None and pf_tgt is not None:
        pf_links_data.append({"source": pf_src, "target": pf_tgt, "value": int(pf_row["value"])})

pf_step_totals = {0: pf_total_users}
for pf_s in range(1, PF_MAX_STEPS + 1):
    pf_v = int(pf_agg.loc[pf_agg["to_step"] == pf_s]["value"].sum())
    if pf_v > 0:
        pf_step_totals[pf_s] = pf_v

pf_method_per_step = {}
for pf_node_item in pf_all_nodes:
    pf_mn = pf_node_item["name"]
    pf_msv = pf_node_item["step"]
    if pf_msv == 0:
        pf_mu = int(pf_entry_counts.get(pf_mn, 0))
    else:
        pf_sl_tmp = pf_agg.loc[(pf_agg["to_step"] == pf_msv) & (pf_agg["to"] == pf_mn)]
        pf_mu = int(pf_sl_tmp["value"].sum()) if not pf_sl_tmp.empty else 0
    if pf_mn not in pf_method_per_step:
        pf_method_per_step[pf_mn] = {}
    pf_method_per_step[pf_mn][str(pf_msv)] = pf_mu

pf_mps_json = json.dumps(pf_method_per_step)
pf_entry_users_json = json.dumps({m: int(pf_entry_counts.get(m, 0)) for m in pf_entry_methods_sorted})


# =========================
# 6) KPI cards
# =========================

def pf_make_kpi(pf_label, pf_tokens, pf_step_total, pf_accent, pf_is_best=False):
    pf_pct = pf_tokens / pf_step_total * 100 if pf_step_total > 0 else 0
    pf_badge = ' <span class="pf-kpi-badge-inline">BEST</span>' if pf_is_best else ""
    return (
        f'<div class="pf-kpi-card" style="--accent:{pf_accent}">'
        f'<div class="pf-kpi-label">{pf_label} <span class="pf-kpi-step">step share</span></div>'
        f'<div class="pf-kpi-value-row"><span class="pf-kpi-value">{pf_pct:.1f}%</span>{pf_badge}</div>'
        f'<div class="pf-kpi-note">{pf_tokens:,} of {pf_step_total:,} step tokens</div>'
        f"</div>"
    )


pf_step_for_kpi = 1
pf_step_total_kpi = int(pf_step_totals.get(pf_step_for_kpi, 0))

pf_step_method_counts = (
    pf_agg.loc[pf_agg["to_step"] == pf_step_for_kpi]
    .groupby("to", as_index=False)["value"]
    .sum()
    .sort_values("value", ascending=False)
)

pf_kpi_top = pf_step_method_counts.head(5).copy()

pf_kpi_data = []
for _, _pf_row in pf_kpi_top.iterrows():
    _pf_method = str(_pf_row["to"])
    _pf_tokens = int(_pf_row["value"])
    _pf_pct = (_pf_tokens / pf_step_total_kpi * 100) if pf_step_total_kpi > 0 else 0
    pf_kpi_data.append((_pf_method, _pf_tokens, _pf_pct))

pf_best_method = max(pf_kpi_data, key=lambda x: x[2])[0] if pf_kpi_data else ""

pf_kpi_html = ""
for pf_m, pf_tokens, _ in pf_kpi_data:
    pf_kpi_html += pf_make_kpi(
        pf_m,
        pf_tokens,
        pf_step_total_kpi,
        PF_PID_COLORS.get(pf_m, XS.C_OTHER),
        pf_is_best=(pf_m == pf_best_method),
    )

# =========================
# 7) Step labels
# =========================

pf_ord = ["1st", "2nd", "3rd", "4th", "5th", "6th", "7th", "8th", "9th", "10th", "11th"]
pf_step_labels = [f"{pf_ord[i]} step" for i in range(PF_MAX_STEPS + 1)]
pf_step_sub = [f"{pf_total_users:,} sessions"] + [
    f"{pf_step_totals.get(pf_s, 0):,} sessions" for pf_s in range(1, PF_MAX_STEPS + 1)
]

pf_subtitle = (
    f"My.Games merchant 799928 · US · only successful payments · {pf_date_range}"
)


# =========================
# 8) Build HTML (Xsolla-themed CSS, d3 Sankey unchanged)
# =========================

_bg = XS.BG_PAGE
_card = XS.BG_PLOT
_primary = XS.TEXT_PRIMARY
_secondary = XS.TEXT_SECONDARY
_border = XS.BORDER
_border_light = XS.BORDER_LIGHT

pf_html = """<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<link href="https://fonts.googleapis.com/css2?family=Open+Sans:wght@400;600;700&family=Roboto:wght@400;500;700&family=Roboto+Mono:wght@400;500&display=swap" rel="stylesheet">
<script src="https://cdnjs.cloudflare.com/ajax/libs/d3/7.8.5/d3.min.js"></script>
<script src="https://cdnjs.cloudflare.com/ajax/libs/d3-sankey/0.12.3/d3-sankey.min.js"></script>
<style>
  *,*::before,*::after{box-sizing:border-box;margin:0;padding:0}
  body{background:__BG__;font-family:__FONT_SANS__;color:__PRIMARY__;padding:24px 16px 36px 32px}
  .pfw{max-width:1700px;margin:0 auto}
  h1{font-size:19px;font-weight:700;color:__PRIMARY__;letter-spacing:-.4px;line-height:1.2;font-family:'Open Sans',sans-serif}
  .pf-sub{font-size:11px;color:__SECONDARY__;margin-top:3px;margin-bottom:16px}

  .pf-kr{display:flex;gap:8px;flex-wrap:wrap;margin-bottom:20px}
  .pf-kpi-card{background:__CARD__;border:1px solid __BORDER__;border-radius:10px;padding:11px 14px;
    min-width:110px;flex:1;position:relative;overflow:hidden;box-shadow:0 1px 4px rgba(0,0,0,.06)}
  .pf-kpi-card::before{content:'';position:absolute;top:0;left:0;right:0;height:2px;background:var(--accent,__ACCENT__)}
  .pf-kpi-label{font-size:10px;color:__SECONDARY__;font-weight:500;line-height:1.3}
  .pf-kpi-step{display:inline-block;background:__SUBTLE__;color:__SECONDARY__;font-size:8px;font-weight:600;
    padding:1px 5px;border-radius:4px;letter-spacing:.05em;margin-left:3px;vertical-align:middle}
  .pf-kpi-value-row{display:flex;align-items:baseline;gap:7px;margin-top:2px}
  .pf-kpi-value{font-size:22px;font-weight:700;color:__PRIMARY__;font-family:'Roboto Mono',monospace;line-height:1.2}
  .pf-kpi-badge-inline{display:inline-block;background:rgba(22,163,74,.12);color:__GOOD__;font-size:9px;font-weight:700;
    padding:2px 7px;border-radius:99px;letter-spacing:.06em;border:1px solid rgba(22,163,74,.35);
    white-space:nowrap;vertical-align:middle}
  .pf-kpi-note{font-size:9px;color:__MUTED__;margin-top:2px}

  .pf-cw{background:__CARD__;border:1px solid __BORDER__;border-radius:14px;padding:20px 16px;box-shadow:0 1px 4px rgba(0,0,0,.06)}
  .pf-cw svg{display:block;width:100%}

  #pf-tt{position:fixed;background:__CARD__;border:1px solid __BORDER__;border-radius:10px;padding:10px 14px;
    font-size:12px;pointer-events:none;opacity:0;transition:opacity .15s;
    box-shadow:0 8px 32px rgba(0,0,0,.12);z-index:1000;max-width:340px;line-height:1.5;font-family:__FONT_SANS2__}
  .pf-ttt{font-weight:600;color:__PRIMARY__;margin-bottom:6px;font-size:12px}
  .pf-ttr{display:flex;justify-content:space-between;gap:16px;color:__SECONDARY__;font-size:11px;padding:1px 0}
  .pf-ttr span:last-child{color:__PRIMARY__;font-family:'Roboto Mono',monospace;font-weight:500}
  .pf-tts{color:__SECONDARY__;margin:7px 0 3px;font-size:10px;border-top:1px solid __BORDER_LIGHT__;padding-top:6px;
    font-weight:600;letter-spacing:.04em;text-transform:uppercase}

  .pf-lr{display:flex;flex-wrap:wrap;gap:10px;margin-top:14px;padding-top:12px;border-top:1px solid __BORDER_LIGHT__}
  .pf-li{display:flex;align-items:center;gap:5px;font-size:11px;color:__SECONDARY__}
  .pf-ld{width:8px;height:8px;border-radius:3px;flex-shrink:0}

  .sankey-link{cursor:pointer}.sankey-link path{transition:opacity .2s}
  .sankey-node{cursor:pointer}.pf-col-h{pointer-events:none}

  .pf-metrics-desc{margin-top:24px;padding-top:20px;border-top:1px solid __BORDER_LIGHT__}
  .pf-md-title{font-size:10px;font-weight:600;color:__SECONDARY__;text-transform:uppercase;
    letter-spacing:.08em;margin-bottom:14px;font-family:'Open Sans',sans-serif}
  .pf-md-grid{display:grid;grid-template-columns:repeat(3,1fr);gap:16px}
  .pf-md-item{background:__SUBTLE__;border:1px solid __BORDER_LIGHT__;border-radius:8px;padding:12px 14px}
  .pf-md-name{font-size:11px;font-weight:700;color:__ACCENT__;margin-bottom:6px;
    font-family:'Open Sans',sans-serif;letter-spacing:-.01em}
  .pf-md-text{font-size:11px;color:__SECONDARY__;line-height:1.65;font-family:'Open Sans',sans-serif}
  .pf-md-text em{font-style:normal;color:__PRIMARY__;font-weight:600}
</style>
</head>
<body>
<div class="pfw">
  <h1>Payment Flow • Only successful payments • My.Games</h1>
  <div class="pf-sub">__SUBTITLE__</div>
  <div class="pf-kr">__KPI_HTML__</div>
  <div class="pf-cw">
    <div id="pf-chart"></div>
    <div class="pf-lr" id="pf-legend"></div>
    <div class="pf-metrics-desc">
      <div class="pf-md-title">Metrics glossary · hover on any flow or node to see values</div>
      <div class="pf-md-grid">
        <div class="pf-md-item">
          <div class="pf-md-name">Preference Rate prev step %</div>
          <div class="pf-md-text">
            Share of sessions on a transition among <em>all sessions leaving the same method node</em> at the previous step.
          </div>
        </div>
        <div class="pf-md-item">
          <div class="pf-md-name">Total Preference Rate %</div>
          <div class="pf-md-text">
            Share on a transition out of <em>all entry sessions</em> for that method at step 0.
          </div>
        </div>
        <div class="pf-md-item">
          <div class="pf-md-name">% of step total</div>
          <div class="pf-md-text">
            Share among <em>all sessions present at that step number</em>, across methods.
          </div>
        </div>
      </div>
    </div>
  </div>
</div>
<div id="pf-tt"></div>

<script>
const pf_ND=__NODES__,pf_LK=__LINKS__,pf_ST=__STEP_TOT__,pf_SL=__STEP_LBL__,
      pf_SS=__STEP_SUB__,pf_CL=__COLORS__,pf_MS=__MAX_STEPS__,
      pf_EU=__ENTRY_USERS__,pf_MPS=__METHOD_PER_STEP__;

const pf_W=Math.max(980,(window.innerWidth||1500)-40),pf_H=532,
      pf_M={top:72,right:130,bottom:24,left:130},
      pf_CW=pf_W-pf_M.left-pf_M.right,pf_CH=pf_H-pf_M.top-pf_M.bottom,
      pf_NW=14,pf_NP=8,pf_MH=4;

const pf_svg=d3.select('#pf-chart').append('svg').attr('viewBox',`0 0 ${pf_W} ${pf_H}`);
const pf_defs=pf_svg.append('defs');
const pf_g=pf_svg.append('g').attr('transform',`translate(${pf_M.left},${pf_M.top})`);

const pf_sk=d3.sankey().nodeId(d=>d.id).nodeAlign(d=>d.step)
  .nodeSort((a,b)=>{
    if(a.step!==b.step)return 0;
    if(a.step===0){
      const ua=pf_EU[a.name]||0,ub=pf_EU[b.name]||0;
      return ub-ua;
    }
    if(a.name==='Other'&&b.name!=='Other')return 1;
    if(b.name==='Other'&&a.name!=='Other')return-1;
    const va=(a.targetLinks||[]).reduce((s,l)=>s+l.value,0)||(a.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
    const vb=(b.targetLinks||[]).reduce((s,l)=>s+l.value,0)||(b.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
    return vb-va;
  }).nodeWidth(pf_NW).nodePadding(pf_NP).iterations(64).extent([[0,0],[pf_CW,pf_CH]]);

const pf_nodes=pf_ND.map(d=>({...d})),pf_links=pf_LK.map(d=>({...d}));
const pf_gr=pf_sk({nodes:pf_nodes,links:pf_links});

// Force exact X position per step so node columns stay within step boundaries.
const pf_stepMax=Math.max(1,pf_MS);
pf_gr.nodes.forEach(n=>{
  const x=(n.step/pf_stepMax)*pf_CW;
  n.x0=x;
  n.x1=x+pf_NW;
});
pf_sk.update(pf_gr);

pf_gr.links.forEach((lk,i)=>{
  const g2=pf_defs.append('linearGradient').attr('id',`pfg${i}`).attr('gradientUnits','userSpaceOnUse')
    .attr('x1',lk.source.x1).attr('x2',lk.target.x0);
  g2.append('stop').attr('offset','0%').attr('stop-color',pf_CL[lk.source.name]||'#94a3b8');
  g2.append('stop').attr('offset','100%').attr('stop-color',pf_CL[lk.target.name]||'#94a3b8');
});

const pf_sx={};
pf_gr.nodes.forEach(n=>{if(!(n.step in pf_sx))pf_sx[n.step]=n.x0+pf_NW/2;});
for(let s=0;s<=pf_MS;s++){
  const cx=pf_sx[s];if(cx===undefined)continue;
  const hg=pf_g.append('g').attr('class','pf-col-h');
  hg.append('rect').attr('x',cx-42).attr('y',-pf_M.top+4).attr('width',84).attr('height',20).attr('rx',5)
    .attr('fill',s===0?'__S12__':'__SUBTLE__').attr('stroke',s===0?'__S7__':'__BORDER__').attr('stroke-width',1);
  hg.append('text').attr('x',cx).attr('y',-pf_M.top+17).attr('text-anchor','middle')
    .attr('fill',s===0?'__S5__':'__SECONDARY__').attr('font-size',10)
    .attr('font-family', "'Open Sans',sans-serif").attr('font-weight',600).attr('letter-spacing','.04em')
    .text((pf_SL[s]||'').toUpperCase());
  hg.append('text').attr('x',cx).attr('y',-pf_M.top+40).attr('text-anchor','middle')
    .attr('fill','__SECONDARY__').attr('font-size',10).attr('font-family', "'Roboto Mono',monospace")
    .text(pf_SS[s]||'');
  if(s>0)pf_g.append('line').attr('x1',cx).attr('y1',-pf_M.top+50).attr('x2',cx).attr('y2',pf_CH)
    .attr('stroke','__BORDER__').attr('stroke-width',1).attr('stroke-dasharray','3,5');
}

const pf_lp=d3.sankeyLinkHorizontal();
const pf_ls=pf_g.append('g').selectAll('.sankey-link').data(pf_gr.links).enter().append('g').attr('class','sankey-link');
pf_ls.append('path').attr('d',pf_lp).attr('fill','none')
  .attr('stroke',(d,i)=>`url(#pfg${i})`).attr('stroke-width',d=>Math.max(1,d.width)).attr('opacity',.28);

const pf_ns=pf_g.append('g').selectAll('.sankey-node').data(pf_gr.nodes).enter().append('g').attr('class','sankey-node')
  .attr('transform',d=>`translate(${d.x0},${d.y0})`);
pf_ns.append('rect').attr('width',pf_NW).attr('height',d=>Math.max(pf_MH,d.y1-d.y0))
  .attr('rx',3).attr('fill',d=>pf_CL[d.name]||'#94a3b8').attr('opacity',.88);
pf_ns.filter(d=>d.step===0||(d.name in pf_EU&&pf_EU[d.name]>0)).append('rect')
  .attr('width',pf_NW).attr('height',d=>Math.max(pf_MH,d.y1-d.y0))
  .attr('rx',3).attr('fill','none')
  .attr('stroke',d=>pf_CL[d.name]||'#94a3b8').attr('stroke-width',1.2).attr('opacity',.5);

const pf_lg=pf_g.append('g');
pf_gr.nodes.forEach(d=>{
  const h=Math.max(pf_MH,d.y1-d.y0),cy=d.y0+h/2;
  const isEntry = d.step===0;
  if(h<6&&!isEntry)return;
  const xpos = isEntry ? d.x0-7 : d.x1+7;
  const an   = isEntry ? 'end'  : 'start';
  const ll   = pf_lg.append('g');
  ll.append('text').attr('x',xpos).attr('y',cy-(h>14?5:0)).attr('text-anchor',an)
    .attr('fill',pf_CL[d.name]||'__SECONDARY__')
    .attr('font-size',isEntry?12:10).attr('font-weight',isEntry?600:400)
    .attr('font-family', "'Open Sans',sans-serif").attr('dominant-baseline',h<=14?'middle':'auto')
    .text(d.name);
  if(isEntry && h>14){
    ll.append('text').attr('x',xpos).attr('y',cy+6).attr('text-anchor',an)
      .attr('fill',pf_CL[d.name]||'__SECONDARY__').attr('font-size',9)
      .attr('font-family', "'Roboto Mono',monospace").attr('font-weight',500)
      .text(`${(pf_EU[d.name]||0).toLocaleString()} entry`);
    return;
  }
  if(h>14){
    const prevU = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step-1)];
    const thisU = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step)]||0;
    if(prevU&&prevU>0){
      const dropPct=((prevU-thisU)/prevU*100).toFixed(1);
      ll.append('text').attr('x',xpos).attr('y',cy+6).attr('text-anchor',an)
        .attr('fill','__BAD__').attr('font-size',8.5)
        .attr('font-family', "'Roboto Mono',monospace").attr('font-weight',500)
        .text(`drop \\u2193 ${dropPct}%`);
    }
  }
});

const pf_tt=d3.select('#pf-tt'),pf_fmt=n=>n.toLocaleString();
const pf_sT=(h,e)=>pf_tt.html(h).style('opacity',1).style('left',(e.clientX+14)+'px').style('top',(e.clientY-10)+'px');
const pf_hT=()=>pf_tt.style('opacity',0);
const pf_mT=e=>pf_tt.style('left',(e.clientX+14)+'px').style('top',(e.clientY-10)+'px');

pf_ls.on('mouseover',function(ev,d){
  pf_ls.selectAll('path').attr('opacity',.05);d3.select(this).select('path').attr('opacity',.75);
  const srcTot   = (d.source.sourceLinks||[]).reduce((s,l)=>s+l.value,0)||(pf_EU[d.source.name]||1);
  const prevPct  = (d.value/srcTot*100).toFixed(1);
  const entryTot = pf_EU[d.source.name]||srcTot;
  const totalPct = (d.value/entryTot*100).toFixed(1);
  const stepTot  = d.source.step===0?(pf_EU[d.source.name]||1):(pf_ST[String(d.source.step)]||1);
  const stepPct  = (d.value/stepTot*100).toFixed(1);
  pf_sT(`
    <div class="pf-ttt">${d.source.name} \\u2192 ${d.target.name}</div>
    <div class="pf-ttr"><span>Departed \\u2192 Arriving</span><span>${pf_fmt(d.value)}</span></div>
    <div class="pf-ttr"><span>Preference Rate prev step %</span><span>${prevPct}%</span></div>
    <div class="pf-ttr"><span>Total Preference Rate %</span><span>${totalPct}%</span></div>
    <div class="pf-ttr"><span>% of step total</span><span>${stepPct}%</span></div>
    <div class="pf-tts">\\u2191 SOURCE (step ${d.source.step+1})</div>
    <div class="pf-ttr"><span>${d.source.name} sessions</span><span>${pf_fmt(srcTot)}</span></div>
  `,ev);
}).on('mousemove',function(ev){pf_mT(ev);}).on('mouseout',function(){pf_ls.selectAll('path').attr('opacity',.28);pf_hT();});

pf_ns.on('mouseover',function(ev,d){
  const cL=new Set(),cN=new Set([d.id]);
  (d.sourceLinks||[]).forEach(l=>{cL.add(l.index);cN.add(l.target.id);});
  (d.targetLinks||[]).forEach(l=>{cL.add(l.index);cN.add(l.source.id);});
  pf_ls.selectAll('path').attr('opacity',(ld,i)=>cL.has(i)?.68:.04);
  pf_ns.selectAll('rect:first-child').attr('opacity',nd=>cN.has(nd.id)?.95:.18);
  const inc     = (d.targetLinks||[]).reduce((s,l)=>s+l.value,0);
  const out     = (d.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
  const stepTot = d.step===0?(pf_EU[d.name]||1):(pf_ST[String(d.step)]||1);
  const prevSame     = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step-1)];
  const thisSame     = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step)]||inc;
  const nextSameLink = (d.sourceLinks||[]).find(l=>l.target.name===d.name);
  const nextSame     = nextSameLink?nextSameLink.value:0;
  const prefPrevPct  = prevSame&&prevSame>0?(thisSame/prevSame*100).toFixed(1):null;
  const sortedOut = [...(d.sourceLinks||[])].sort((a,b)=>b.value-a.value).slice(0,5);
  const nextHtml  = out>0?sortedOut.map(l=>
    `<div class="pf-ttr"><span>\\u2192 ${l.target.name}</span><span>${pf_fmt(l.value)} (${(l.value/out*100).toFixed(1)}%)</span></div>`
  ).join(''):'';
  const tgtInc  = (d.targetLinks||[]).reduce((s,l)=>s+l.value,0);
  const sortedSrc = [...(d.targetLinks||[])].sort((a,b)=>b.value-a.value).slice(0,3);
  const srcHtml = tgtInc>0?sortedSrc.map(l=>
    `<div class="pf-ttr"><span>\\u2190 ${l.source.name}</span><span>${pf_fmt(l.value)} (${(l.value/tgtInc*100).toFixed(1)}%)</span></div>`
  ).join(''):'';
  if(d.step===0){
    pf_sT(`
      <div class="pf-ttt">${d.name} \\u00b7 Step 1</div>
      <div class="pf-ttr"><span>Entry sessions</span><span>${pf_fmt(pf_EU[d.name]||0)}</span></div>
      <div class="pf-ttr"><span>With next step</span><span>${pf_fmt(out)}</span></div>
      ${nextHtml?`<div class="pf-tts">\\u2192 Next</div>${nextHtml}`:''}
    `,ev);
    return;
  }
  pf_sT(`
    <div class="pf-ttt">${d.name} \\u00b7 Step ${d.step+1}</div>
    ${prevSame?`<div class="pf-ttr"><span>${d.name} at prev step</span><span>${pf_fmt(prevSame)}</span></div>`:''}
    <div class="pf-ttr"><span>${d.name} at this step</span><span>${pf_fmt(thisSame)}</span></div>
    ${prefPrevPct?`<div class="pf-ttr"><span>Preference Rate prev step %</span><span>${prefPrevPct}%</span></div>`:''}
    ${nextSame>0?`<div class="pf-ttr"><span>${d.name} at next step</span><span>${pf_fmt(nextSame)}</span></div>`:''}
    <div class="pf-ttr"><span>Share of step</span><span>${(Math.max(inc,out)/stepTot*100).toFixed(1)}%</span></div>
    ${nextHtml?`<div class="pf-tts">\\u2192 Next</div>${nextHtml}`:''}
    ${srcHtml?`<div class="pf-tts">\\u2190 Top sources</div>${srcHtml}`:''}
  `,ev);
}).on('mousemove',function(ev){pf_mT(ev);})
  .on('mouseout',function(){pf_ls.selectAll('path').attr('opacity',.28);pf_ns.selectAll('rect:first-child').attr('opacity',.88);pf_hT();});

const pf_le=document.getElementById('pf-legend'),pf_un=new Set(pf_gr.nodes.map(n=>n.name));
Object.entries(pf_CL).filter(([n])=>pf_un.has(n)).forEach(([n,c])=>{
  const it=document.createElement('div');it.className='pf-li';
  it.innerHTML=`<div class="pf-ld" style="background:${c}"></div>${n}`;pf_le.appendChild(it);
});

pf_ls.selectAll('path')
  .attr('stroke-dasharray',function(){return this.getTotalLength()+' '+this.getTotalLength();})
  .attr('stroke-dashoffset',function(){return this.getTotalLength();})
  .transition().duration(1400).delay((d,i)=>50+i*4).ease(d3.easeCubicOut)
  .attr('stroke-dashoffset',0).on('end',function(){d3.select(this).attr('stroke-dasharray',null);});
pf_ns.attr('opacity',0).transition().duration(400).delay(d=>d.step*70+120).attr('opacity',1);
</script></body></html>"""

pf_html = (
    pf_html.replace("__BG__", _bg)
    .replace("__CARD__", _card)
    .replace("__PRIMARY__", _primary)
    .replace("__SECONDARY__", _secondary)
    .replace("__BORDER__", _border)
    .replace("__BORDER_LIGHT__", _border_light)
    .replace("__SUBTLE__", XS.BG_SUBTLE)
    .replace("__S12__", XS.S12)
    .replace("__S7__", XS.S7)
    .replace("__S5__", XS.S5)
    .replace("__ACCENT__", XS.S7)
    .replace("__MUTED__", XS.TEXT_HINT)
    .replace("__GOOD__", XS.GOOD)
    .replace("__BAD__", XS.BAD)
    .replace("__FONT_SANS__", "'Open Sans', 'Roboto', sans-serif")
    .replace("__FONT_SANS2__", "'Open Sans', 'Roboto', sans-serif")
    .replace("__SUBTITLE__", pf_subtitle)
    .replace("__KPI_HTML__", pf_kpi_html)
    .replace("__NODES__", json.dumps(pf_all_nodes))
    .replace("__LINKS__", json.dumps(pf_links_data))
    .replace("__STEP_TOT__", json.dumps({str(k): v for k, v in pf_step_totals.items()}))
    .replace("__STEP_LBL__", json.dumps(pf_step_labels))
    .replace("__STEP_SUB__", json.dumps(pf_step_sub))
    .replace("__COLORS__", json.dumps(PF_PID_COLORS))
    .replace("__MAX_STEPS__", str(PF_MAX_STEPS))
    .replace("__ENTRY_USERS__", pf_entry_users_json)
    .replace("__METHOD_PER_STEP__", pf_mps_json)
)


# =========================
# 9) Confluence export
# =========================

pf_conf = (
    pf_html.replace("background:" + _bg + ";", "background:transparent;")
    .replace("padding:24px 16px 36px 32px;", "padding:8px 4px 16px;")
    .replace("max-width:1700px;", "max-width:100%;")
)

pf_export = (
    '<div style="font-family:sans-serif;margin-top:20px;border-radius:12px;'
    'border:1px solid #e2e8f0;overflow:hidden;box-shadow:0 1px 6px rgba(0,0,0,.07);">'
    '<button onclick="var b=document.getElementById(\'pfcb\');var a=document.getElementById(\'pfca\');'
    'var o=b.style.display!==\'none\';b.style.display=o?\'none\':\'block\';'
    'a.style.transform=o?\'rotate(0deg)\':\'rotate(180deg)\';"'
    ' style="width:100%;background:#f8fafc;border:none;border-bottom:1px solid #e2e8f0;'
    'padding:14px 20px;display:flex;align-items:center;justify-content:space-between;cursor:pointer;">'
    '<div><div style="font-size:13px;font-weight:600;color:#0f172a;">Confluence HTML'
    '<span style="background:#EBFCFF;color:#1B8398;font-size:10px;font-weight:600;'
    'padding:2px 8px;border-radius:99px;margin-left:6px;">READY TO PASTE</span></div>'
    '<div style="font-size:11px;color:#94a3b8;margin-top:1px;">'
    'Copy and paste into HTML Macro (height: 1000px)</div></div>'
    '<span id="pfca" style="transition:transform .25s;font-size:18px;color:#94a3b8;">&#8964;</span>'
    '</button>'
    '<div id="pfcb" style="display:none;background:#fff;">'
    '<div style="position:relative;background:#f8fafc;border-bottom:1px solid #f1f5f9;">'
    '<textarea id="pfcc" readonly style="width:100%;height:140px;font-family:monospace;font-size:11px;'
    'color:#475569;background:transparent;border:none;padding:14px 20px;'
    'resize:none;outline:none;line-height:1.6;box-sizing:border-box;">'
    + pf_conf.replace("'", "&#39;")
    + '</textarea>'
    + '<div id="pfck" style="position:absolute;bottom:8px;right:14px;font-size:10px;'
    'color:#cbd5e1;font-family:monospace;pointer-events:none;"></div></div>'
    '<div style="padding:12px 20px;display:flex;align-items:center;gap:12px;">'
    '<button id="pfcpb" onclick="pfCopy()" style="background:#31CAE8;color:#fff;border:none;'
    'border-radius:8px;padding:9px 20px;font-size:13px;font-weight:600;cursor:pointer;transition:background .2s;">'
    'Copy to clipboard</button>'
    '<span id="pfcps" style="font-size:12px;color:#94a3b8;"></span>'
    '</div></div></div>'
    '<script>'
    'var pfta=document.getElementById("pfcc");'
    'document.getElementById("pfck").textContent=(pfta.value.length/1024).toFixed(1)+" KB";'
    'function pfCopy(){'
    'navigator.clipboard.writeText(pfta.value).then(function(){'
    'var btn=document.getElementById("pfcpb"),st=document.getElementById("pfcps");'
    'btn.style.background="#16a34a";btn.textContent="Copied!";'
    'st.textContent="Copied";st.style.color="#16a34a";'
    'setTimeout(function(){btn.style.background="#31CAE8";btn.textContent="Copy to clipboard";st.textContent="";},3000);'
    '}).catch(function(){pfta.select();document.execCommand("copy");});}'
    '</script>'
)

# Jupyter sanitizes inline <script> in HTML(); serve the same file over HTTP and embed via IFrame (D3 works).
import functools
import http.server
import socketserver
import threading
import webbrowser

_pf_preview_path = (Path.cwd() / "payment_flow_sankey_preview.html").resolve()
_pf_preview_path.write_text(pf_html, encoding="utf-8")

class _PfTCPServer(socketserver.TCPServer):
    allow_reuse_address = True


class _PfQuietHandler(http.server.SimpleHTTPRequestHandler):
    def log_message(self, format, *args):
        return


_pf_root = _pf_preview_path.parent
_pf_httpd = None
_pf_port = None
for _pf_try_port in range(8765, 8795):
    try:
        _pf_handler = functools.partial(
            _PfQuietHandler,
            directory=str(_pf_root),
        )
        _pf_httpd = _PfTCPServer(("127.0.0.1", _pf_try_port), _pf_handler)
        _pf_port = _pf_try_port
        break
    except OSError:
        continue

if _pf_httpd is not None and _pf_port is not None:
    threading.Thread(target=_pf_httpd.serve_forever, daemon=True).start()
    _pf_iframe_url = f"http://127.0.0.1:{_pf_port}/{_pf_preview_path.name}"
    print("Embedded chart (HTTP iframe, scripts allowed):", _pf_iframe_url)
    display(IFrame(src=_pf_iframe_url, width="100%", height=620))
else:
    print("Could not start local HTTP preview; open the HTML file or use the VS Code task.")















```

```python
# Payment Flow (df_payment_flow_my_games) — Template 1 structure, Xsolla palette, session step logic

import gc
import json
import sys
from pathlib import Path

import pandas as pd
from IPython.display import HTML, IFrame, display

_cwd = Path.cwd()
_xsolla_root = _cwd if (_cwd / "utils" / "style.py").is_file() else _cwd.parent
if str(_xsolla_root) not in sys.path:
    sys.path.insert(0, str(_xsolla_root))
from utils.style import XS

# =========================
# 1) Parameters
# =========================

PF_MAX_STEPS = 4
PF_TOP_METHODS = 10
PF_DATE_COL = "internal_date"

_xs_pal = [XS.S3, XS.S5, XS.S7, XS.S9, XS.S6, XS.S8, XS.S4, XS.S10]


def pf_pid_label(raw):
    if pd.isna(raw):
        return "Unknown"
    s = str(raw).strip()
    parts = s.split(".", 1)
    return parts[1].strip() if len(parts) > 1 else s


def pf_row_flow_label(row):
    if int(row["sankey_step"]) == 0:
        return pf_pid_label(row["pid"])
    return pf_pid_label(row["bill_pid"])


def pf_build_colors(names):
    out = {"Other": XS.C_OTHER}
    names = sorted(n for n in names if n and n != "Other")
    for i, n in enumerate(names):
        out[n] = _xs_pal[i % len(_xs_pal)]
    return out


# =========================
# 2) Data preparation
# =========================

pf_df = df_payment_flow_my_games.copy()
pf_df = pf_df.dropna(subset=["pid"]).copy()
pf_df["internal_time"] = pd.to_datetime(pf_df["internal_time"], errors="coerce")
pf_df = pf_df.sort_values(["user_access_token", "steps"]).reset_index(drop=True)
pf_df["sankey_step"] = pf_df.groupby("user_access_token", sort=False).cumcount().astype("int64")
pf_df["flow_label"] = pf_df.apply(pf_row_flow_label, axis=1)
pf_df["pid_label"] = pf_df["flow_label"]
gc.collect()

try:
    pf_date_range = f"{pf_df[PF_DATE_COL].min()} – {pf_df[PF_DATE_COL].max()}"
except Exception:
    pf_date_range = ""

pf_df_entry = (
    pf_df.loc[pf_df["sankey_step"] == 0, ["user_access_token", "pid_label"]]
    .drop_duplicates(subset=["user_access_token"])
    .rename(columns={"pid_label": "entry_method"})
    .copy()
)

# Entry step labels: keep top-5 methods, group all others into Other.
pf_entry_counts_all = pf_df_entry["entry_method"].value_counts()
pf_entry_top5 = set(pf_entry_counts_all.head(5).index)
pf_entry_mandatory = {"Bank card"}
pf_entry_allowed = pf_entry_mandatory
pf_df_entry = pf_df_entry.loc[pf_df_entry["entry_method"].isin(pf_entry_allowed)].copy()

pf_entry_counts = pf_df_entry["entry_method"].value_counts()
pf_entry_methods_sorted = pf_entry_counts.index.tolist()
pf_total_users = len(pf_df_entry)

pf_df_after = pf_df.loc[(pf_df["sankey_step"] >= 1) & (pf_df["done_just"] == 1) & (pf_df["bill_pid"].notna())].copy()
pf_df_after = pf_df_after.merge(
    pf_df_entry[["user_access_token", "entry_method"]], on="user_access_token", how="inner"
)
pf_df_after.sort_values(["user_access_token", "sankey_step"], inplace=True)
pf_df_after.reset_index(drop=True, inplace=True)
pf_df_after["step"] = pf_df_after["sankey_step"].astype("int16")
pf_df_after = pf_df_after.loc[pf_df_after["step"] <= PF_MAX_STEPS].copy()
gc.collect()


# =========================
# 3) Vectorized transitions
# =========================

pf_df_step0 = pf_df_entry[["user_access_token", "entry_method"]].rename(
    columns={"entry_method": "pid_label"}
).copy()
pf_df_step0["step"] = 0

pf_df_all_steps = pd.concat(
    [pf_df_step0[["user_access_token", "step", "pid_label"]], pf_df_after[["user_access_token", "step", "pid_label"]]],
    ignore_index=True,
)
pf_df_all_steps.sort_values(["user_access_token", "step"], inplace=True)
pf_df_all_steps.reset_index(drop=True, inplace=True)

pf_df_all_steps["next_pid"] = pf_df_all_steps.groupby("user_access_token")["pid_label"].shift(-1)
pf_df_all_steps["next_step"] = pf_df_all_steps.groupby("user_access_token")["step"].shift(-1)

pf_df_trans = pf_df_all_steps.dropna(subset=["next_pid", "next_step"]).copy()
pf_df_trans = pf_df_trans.loc[pf_df_trans["next_step"] == pf_df_trans["step"] + 1].copy()
pf_df_trans = pf_df_trans.rename(
    columns={
        "step": "from_step",
        "pid_label": "from",
        "next_step": "to_step",
        "next_pid": "to",
    }
)
pf_df_trans["to_step"] = pf_df_trans["to_step"].astype("int16")

del pf_df_all_steps, pf_df_after, pf_df_step0, pf_df
gc.collect()


# =========================
# 4) Top-N per step + remap Other
# =========================

pf_top_per_step = {}
for pf_step in range(1, PF_MAX_STEPS + 1):
    pf_sl = pf_df_trans.loc[pf_df_trans["to_step"] == pf_step]
    if pf_sl.empty:
        pf_top_per_step[pf_step] = set(pf_entry_methods_sorted)
        continue
    pf_top_set = set(pf_sl["to"].value_counts().head(PF_TOP_METHODS).index)
    for pf_em in pf_entry_methods_sorted:
        pf_top_set.add(pf_em)
    pf_top_per_step[pf_step] = pf_top_set


def pf_remap(pf_lbl, pf_stp):
    if pf_stp == 0:
        return pf_lbl
    return pf_lbl if pf_lbl in pf_top_per_step.get(pf_stp, set()) else "Other"


pf_df_trans["from_r"] = pf_df_trans.apply(lambda r: pf_remap(r["from"], int(r["from_step"])), axis=1)
pf_df_trans["to_r"] = pf_df_trans.apply(lambda r: pf_remap(r["to"], int(r["to_step"])), axis=1)

pf_agg = pf_df_trans.groupby(["from_step", "to_step", "from_r", "to_r"]).size().reset_index(name="value")
pf_agg.columns = ["from_step", "to_step", "from", "to", "value"]
del pf_df_trans
gc.collect()


# =========================
# 5) Sankey nodes & links
# =========================

PF_PID_COLORS = pf_build_colors(
    set(pf_agg["from"]).union(pf_agg["to"]).union(pf_entry_methods_sorted)
)

pf_node_lookup, pf_all_nodes, pf_nid = {}, [], 0

for pf_s in range(PF_MAX_STEPS + 1):
    if pf_s == 0:
        pf_pids = [m for m in pf_entry_methods_sorted if m in pf_df_entry["entry_method"].values]
    else:
        pf_sl = pf_agg.loc[pf_agg["to_step"] == pf_s]
        if pf_sl.empty:
            continue
        pf_vol = pf_sl.groupby("to")["value"].sum().sort_values(ascending=False)
        pf_has_other = "Other" in pf_vol.index
        pf_entry_in_vol = [m for m in pf_entry_methods_sorted if m in pf_vol.index]
        pf_middle = [p for p in pf_vol.index if p not in pf_entry_methods_sorted and p != "Other"]
        pf_pids = pf_entry_in_vol + pf_middle + (["Other"] if pf_has_other else [])

    for pf_pid in pf_pids:
        pf_node_lookup[(pf_s, pf_pid)] = pf_nid
        pf_all_nodes.append({"id": pf_nid, "name": pf_pid, "step": pf_s})
        pf_nid += 1

pf_links_data = []
for _, pf_row in pf_agg.iterrows():
    pf_src = pf_node_lookup.get((int(pf_row["from_step"]), pf_row["from"]))
    pf_tgt = pf_node_lookup.get((int(pf_row["to_step"]), pf_row["to"]))
    if pf_src is not None and pf_tgt is not None:
        pf_links_data.append({"source": pf_src, "target": pf_tgt, "value": int(pf_row["value"])})

pf_step_totals = {0: pf_total_users}
for pf_s in range(1, PF_MAX_STEPS + 1):
    pf_v = int(pf_agg.loc[pf_agg["to_step"] == pf_s]["value"].sum())
    if pf_v > 0:
        pf_step_totals[pf_s] = pf_v

pf_method_per_step = {}
for pf_node_item in pf_all_nodes:
    pf_mn = pf_node_item["name"]
    pf_msv = pf_node_item["step"]
    if pf_msv == 0:
        pf_mu = int(pf_entry_counts.get(pf_mn, 0))
    else:
        pf_sl_tmp = pf_agg.loc[(pf_agg["to_step"] == pf_msv) & (pf_agg["to"] == pf_mn)]
        pf_mu = int(pf_sl_tmp["value"].sum()) if not pf_sl_tmp.empty else 0
    if pf_mn not in pf_method_per_step:
        pf_method_per_step[pf_mn] = {}
    pf_method_per_step[pf_mn][str(pf_msv)] = pf_mu

pf_mps_json = json.dumps(pf_method_per_step)
pf_entry_users_json = json.dumps({m: int(pf_entry_counts.get(m, 0)) for m in pf_entry_methods_sorted})


# =========================
# 6) KPI cards
# =========================

def pf_make_kpi(pf_label, pf_tokens, pf_step_total, pf_accent, pf_is_best=False):
    pf_pct = pf_tokens / pf_step_total * 100 if pf_step_total > 0 else 0
    pf_badge = ' <span class="pf-kpi-badge-inline">BEST</span>' if pf_is_best else ""
    return (
        f'<div class="pf-kpi-card" style="--accent:{pf_accent}">'
        f'<div class="pf-kpi-label">{pf_label} <span class="pf-kpi-step">2nd step share</span></div>'
        f'<div class="pf-kpi-value-row"><span class="pf-kpi-value">{pf_pct:.1f}%</span>{pf_badge}</div>'
        f'<div class="pf-kpi-note">{pf_tokens:,} of {pf_step_total:,} step tokens</div>'
        f"</div>"
    )


pf_step_for_kpi = 1
pf_step_total_kpi = int(pf_step_totals.get(pf_step_for_kpi, 0))

pf_step_method_counts = (
    pf_agg.loc[pf_agg["to_step"] == pf_step_for_kpi]
    .groupby("to", as_index=False)["value"]
    .sum()
    .sort_values("value", ascending=False)
)

pf_kpi_top = pf_step_method_counts.head(5).copy()

pf_kpi_data = []
for _, _pf_row in pf_kpi_top.iterrows():
    _pf_method = str(_pf_row["to"])
    _pf_tokens = int(_pf_row["value"])
    _pf_pct = (_pf_tokens / pf_step_total_kpi * 100) if pf_step_total_kpi > 0 else 0
    pf_kpi_data.append((_pf_method, _pf_tokens, _pf_pct))

pf_best_method = max(pf_kpi_data, key=lambda x: x[2])[0] if pf_kpi_data else ""

pf_kpi_html = ""
for pf_m, pf_tokens, _ in pf_kpi_data:
    pf_kpi_html += pf_make_kpi(
        pf_m,
        pf_tokens,
        pf_step_total_kpi,
        PF_PID_COLORS.get(pf_m, XS.C_OTHER),
        pf_is_best=(pf_m == pf_best_method),
    )

# =========================
# 7) Step labels
# =========================

pf_ord = ["1st", "2nd", "3rd", "4th", "5th", "6th", "7th", "8th", "9th", "10th", "11th"]
pf_step_labels = [f"{pf_ord[i]} step" for i in range(PF_MAX_STEPS + 1)]
pf_step2_share_pct = (pf_step_totals.get(1, 0) / pf_total_users * 100) if pf_total_users else 0.0
pf_step_sub = [f"{pf_total_users:,} sessions"] + [
    (f"{pf_step2_share_pct:.1f}% of 1st step" if pf_s == 1 else f"{pf_step_totals.get(pf_s, 0):,} sessions")
    for pf_s in range(1, PF_MAX_STEPS + 1)
]

pf_subtitle = (
    f"My.Games merchant 799928 · US · only successful payments · {pf_date_range}"
)


# =========================
# 8) Build HTML (Xsolla-themed CSS, d3 Sankey unchanged)
# =========================

_bg = XS.BG_PAGE
_card = XS.BG_PLOT
_primary = XS.TEXT_PRIMARY
_secondary = XS.TEXT_SECONDARY
_border = XS.BORDER
_border_light = XS.BORDER_LIGHT

pf_html = """<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<link href="https://fonts.googleapis.com/css2?family=Open+Sans:wght@400;600;700&family=Roboto:wght@400;500;700&family=Roboto+Mono:wght@400;500&display=swap" rel="stylesheet">
<script src="https://cdnjs.cloudflare.com/ajax/libs/d3/7.8.5/d3.min.js"></script>
<script src="https://cdnjs.cloudflare.com/ajax/libs/d3-sankey/0.12.3/d3-sankey.min.js"></script>
<script src="https://cdnjs.cloudflare.com/ajax/libs/html2canvas/1.4.1/html2canvas.min.js"></script>
<style>
  *,*::before,*::after{box-sizing:border-box;margin:0;padding:0}
  body{background:__BG__;font-family:__FONT_SANS__;color:__PRIMARY__;padding:24px 16px 36px 32px}
  .pfw{max-width:1700px;margin:0 auto}
  h1{font-size:19px;font-weight:700;color:__PRIMARY__;letter-spacing:-.4px;line-height:1.2;font-family:'Open Sans',sans-serif}
  .pf-sub{font-size:11px;color:__SECONDARY__;margin-top:3px;margin-bottom:16px}

  .pf-kr{display:flex;gap:8px;flex-wrap:wrap;margin-bottom:20px}
  .pf-kpi-card{background:__CARD__;border:1px solid __BORDER__;border-radius:10px;padding:11px 14px;
    min-width:110px;flex:1;position:relative;overflow:hidden;box-shadow:0 1px 4px rgba(0,0,0,.06)}
  .pf-kpi-card::before{content:'';position:absolute;top:0;left:0;right:0;height:2px;background:var(--accent,__ACCENT__)}
  .pf-kpi-label{font-size:10px;color:__SECONDARY__;font-weight:500;line-height:1.3}
  .pf-kpi-step{display:inline-block;background:__SUBTLE__;color:__SECONDARY__;font-size:8px;font-weight:600;
    padding:1px 5px;border-radius:4px;letter-spacing:.05em;margin-left:3px;vertical-align:middle}
  .pf-kpi-value-row{display:flex;align-items:baseline;gap:7px;margin-top:2px}
  .pf-kpi-value{font-size:22px;font-weight:700;color:__PRIMARY__;font-family:'Roboto Mono',monospace;line-height:1.2}
  .pf-kpi-badge-inline{display:inline-block;background:rgba(22,163,74,.12);color:__GOOD__;font-size:9px;font-weight:700;
    padding:2px 7px;border-radius:99px;letter-spacing:.06em;border:1px solid rgba(22,163,74,.35);
    white-space:nowrap;vertical-align:middle}
  .pf-kpi-note{font-size:9px;color:__MUTED__;margin-top:2px}

  .pf-cw{background:__CARD__;border:1px solid __BORDER__;border-radius:14px;padding:20px 16px;box-shadow:0 1px 4px rgba(0,0,0,.06)}
  .pf-cw svg{display:block;width:100%}

  #pf-tt{position:fixed;background:__CARD__;border:1px solid __BORDER__;border-radius:10px;padding:10px 14px;
    font-size:12px;pointer-events:none;opacity:0;transition:opacity .15s;
    box-shadow:0 8px 32px rgba(0,0,0,.12);z-index:1000;max-width:340px;line-height:1.5;font-family:__FONT_SANS2__}
  .pf-ttt{font-weight:600;color:__PRIMARY__;margin-bottom:6px;font-size:12px}
  .pf-ttr{display:flex;justify-content:space-between;gap:16px;color:__SECONDARY__;font-size:11px;padding:1px 0}
  .pf-ttr span:last-child{color:__PRIMARY__;font-family:'Roboto Mono',monospace;font-weight:500}
  .pf-tts{color:__SECONDARY__;margin:7px 0 3px;font-size:10px;border-top:1px solid __BORDER_LIGHT__;padding-top:6px;
    font-weight:600;letter-spacing:.04em;text-transform:uppercase}

  .pf-lr{display:flex;flex-wrap:wrap;gap:10px;margin-top:14px;padding-top:12px;border-top:1px solid __BORDER_LIGHT__}
  .pf-li{display:flex;align-items:center;gap:5px;font-size:11px;color:__SECONDARY__}
  .pf-ld{width:8px;height:8px;border-radius:3px;flex-shrink:0}
  .pf-save-bottom{display:flex;align-items:center;justify-content:flex-end;gap:10px;margin-top:12px}
  .pf-save-btn{background:__S7__;color:#fff;border:none;border-radius:8px;padding:7px 12px;font-size:12px;font-weight:600;cursor:pointer}
  .pf-save-status{font-size:11px;color:__SECONDARY__}

  .sankey-link{cursor:pointer}.sankey-link path{transition:opacity .2s}
  .sankey-node{cursor:pointer}.pf-col-h{pointer-events:none}

  .pf-metrics-desc{margin-top:24px;padding-top:20px;border-top:1px solid __BORDER_LIGHT__}
  .pf-md-title{font-size:10px;font-weight:600;color:__SECONDARY__;text-transform:uppercase;
    letter-spacing:.08em;margin-bottom:14px;font-family:'Open Sans',sans-serif}
  .pf-md-grid{display:grid;grid-template-columns:repeat(2,1fr);gap:16px}
  .pf-md-item{background:__SUBTLE__;border:1px solid __BORDER_LIGHT__;border-radius:8px;padding:12px 14px}
  .pf-md-name{font-size:11px;font-weight:700;color:__ACCENT__;margin-bottom:6px;
    font-family:'Open Sans',sans-serif;letter-spacing:-.01em}
  .pf-md-text{font-size:11px;color:__SECONDARY__;line-height:1.65;font-family:'Open Sans',sans-serif}
  .pf-md-text em{font-style:normal;color:__PRIMARY__;font-weight:600}
</style>
</head>
<body>
<div class="pfw">
  <h1>Payment Flow • Bank cards flow • Only successful payments • My.Games</h1>
  <div class="pf-sub">__SUBTITLE__</div>
  <div class="pf-kr">__KPI_HTML__</div>
  <div class="pf-cw">
    <div id="pf-chart"></div>
    <div class="pf-lr" id="pf-legend"></div>
    <div class="pf-metrics-desc">
      <div class="pf-md-title">Metrics glossary</div>
      <div class="pf-md-grid">
        <div class="pf-md-item">
          <div class="pf-md-name">1st step metric</div>
          <div class="pf-md-text">
            Number of sessions with interaction with the <em>Bank card</em> payment method.
          </div>
        </div>
        <div class="pf-md-item">
          <div class="pf-md-name">2nd+ steps metric</div>
          <div class="pf-md-text">
            The second and subsequent steps show only sessions with <em>successful payments</em>.
          </div>
        </div>
      </div>
    </div>
  </div>
  <div class="pf-save-bottom"><button id="pf-save-svg" class="pf-save-btn">Save as SVG</button><span id="pf-save-status" class="pf-save-status"></span></div>
</div>
<div id="pf-tt"></div>

<script>
const pf_ND=__NODES__,pf_LK=__LINKS__,pf_ST=__STEP_TOT__,pf_SL=__STEP_LBL__,
      pf_SS=__STEP_SUB__,pf_CL=__COLORS__,pf_MS=__MAX_STEPS__,
      pf_EU=__ENTRY_USERS__,pf_MPS=__METHOD_PER_STEP__;

const pf_W=Math.max(980,(window.innerWidth||1500)-40),pf_H=426,
      pf_M={top:72,right:130,bottom:24,left:130},
      pf_CW=pf_W-pf_M.left-pf_M.right,pf_CH=pf_H-pf_M.top-pf_M.bottom,
      pf_NW=14,pf_NP=8,pf_MH=4;

const pf_svg=d3.select('#pf-chart').append('svg').attr('viewBox',`0 0 ${pf_W} ${pf_H}`);
const pf_defs=pf_svg.append('defs');
const pf_g=pf_svg.append('g').attr('transform',`translate(${pf_M.left},${pf_M.top})`);

const pf_sk=d3.sankey().nodeId(d=>d.id).nodeAlign(d=>d.step)
  .nodeSort((a,b)=>{
    if(a.step!==b.step)return 0;
    if(a.step===0){
      const ua=pf_EU[a.name]||0,ub=pf_EU[b.name]||0;
      return ub-ua;
    }
    if(a.name==='Other'&&b.name!=='Other')return 1;
    if(b.name==='Other'&&a.name!=='Other')return-1;
    const va=(a.targetLinks||[]).reduce((s,l)=>s+l.value,0)||(a.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
    const vb=(b.targetLinks||[]).reduce((s,l)=>s+l.value,0)||(b.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
    return vb-va;
  }).nodeWidth(pf_NW).nodePadding(pf_NP).iterations(64).extent([[0,0],[pf_CW,pf_CH]]);

const pf_nodes=pf_ND.map(d=>({...d})),pf_links=pf_LK.map(d=>({...d}));
const pf_gr=pf_sk({nodes:pf_nodes,links:pf_links});

// Force exact X position per step so node columns stay within step boundaries.
const pf_stepMax=Math.max(1,pf_MS);
pf_gr.nodes.forEach(n=>{
  const x=(n.step/pf_stepMax)*pf_CW;
  n.x0=x;
  n.x1=x+pf_NW;
});
pf_sk.update(pf_gr);

pf_gr.links.forEach((lk,i)=>{
  const g2=pf_defs.append('linearGradient').attr('id',`pfg${i}`).attr('gradientUnits','userSpaceOnUse')
    .attr('x1',lk.source.x1).attr('x2',lk.target.x0);
  g2.append('stop').attr('offset','0%').attr('stop-color',pf_CL[lk.source.name]||'#94a3b8');
  g2.append('stop').attr('offset','100%').attr('stop-color',pf_CL[lk.target.name]||'#94a3b8');
});

const pf_sx={};
pf_gr.nodes.forEach(n=>{if(!(n.step in pf_sx))pf_sx[n.step]=n.x0+pf_NW/2;});
for(let s=0;s<=pf_MS;s++){
  const cx=pf_sx[s];if(cx===undefined)continue;
  const hg=pf_g.append('g').attr('class','pf-col-h');
  hg.append('rect').attr('x',cx-42).attr('y',-pf_M.top+4).attr('width',84).attr('height',20).attr('rx',5)
    .attr('fill',s===0?'__S12__':'__SUBTLE__').attr('stroke',s===0?'__S7__':'__BORDER__').attr('stroke-width',1);
  hg.append('text').attr('x',cx).attr('y',-pf_M.top+17).attr('text-anchor','middle')
    .attr('fill',s===0?'__S5__':'__SECONDARY__').attr('font-size',10)
    .attr('font-family', "'Open Sans',sans-serif").attr('font-weight',600).attr('letter-spacing','.04em')
    .text((pf_SL[s]||'').toUpperCase());
  hg.append('text').attr('x',cx).attr('y',-pf_M.top+40).attr('text-anchor','middle')
    .attr('fill','__SECONDARY__').attr('font-size',10).attr('font-family', "'Roboto Mono',monospace")
    .text(pf_SS[s]||'');
  if(s>0)pf_g.append('line').attr('x1',cx).attr('y1',-pf_M.top+50).attr('x2',cx).attr('y2',pf_CH)
    .attr('stroke','__BORDER__').attr('stroke-width',1).attr('stroke-dasharray','3,5');
}

const pf_lp=d3.sankeyLinkHorizontal();
const pf_ls=pf_g.append('g').selectAll('.sankey-link').data(pf_gr.links).enter().append('g').attr('class','sankey-link');
pf_ls.append('path').attr('d',pf_lp).attr('fill','none')
  .attr('stroke',(d,i)=>`url(#pfg${i})`).attr('stroke-width',d=>Math.max(1,d.width)).attr('opacity',.28);

const pf_ns=pf_g.append('g').selectAll('.sankey-node').data(pf_gr.nodes).enter().append('g').attr('class','sankey-node')
  .attr('transform',d=>`translate(${d.x0},${d.y0})`);
pf_ns.append('rect').attr('width',pf_NW).attr('height',d=>Math.max(pf_MH,d.y1-d.y0))
  .attr('rx',3).attr('fill',d=>pf_CL[d.name]||'#94a3b8').attr('opacity',.88);
pf_ns.filter(d=>d.step===0||(d.name in pf_EU&&pf_EU[d.name]>0)).append('rect')
  .attr('width',pf_NW).attr('height',d=>Math.max(pf_MH,d.y1-d.y0))
  .attr('rx',3).attr('fill','none')
  .attr('stroke',d=>pf_CL[d.name]||'#94a3b8').attr('stroke-width',1.2).attr('opacity',.5);

const pf_lg=pf_g.append('g');
pf_gr.nodes.forEach(d=>{
  const h=Math.max(pf_MH,d.y1-d.y0),cy=d.y0+h/2;
  const isEntry = d.step===0;
  if(h<6&&!isEntry)return;
  const xpos = isEntry ? d.x0-7 : d.x1+7;
  const an   = isEntry ? 'end'  : 'start';
  const ll   = pf_lg.append('g');
  ll.append('text').attr('x',xpos).attr('y',cy-(h>14?5:0)).attr('text-anchor',an)
    .attr('fill',pf_CL[d.name]||'__SECONDARY__')
    .attr('font-size',isEntry?12:10).attr('font-weight',isEntry?600:400)
    .attr('font-family', "'Open Sans',sans-serif").attr('dominant-baseline',h<=14?'middle':'auto')
    .text(d.name);
  if(isEntry && h>14){
    ll.append('text').attr('x',xpos).attr('y',cy+6).attr('text-anchor',an)
      .attr('fill',pf_CL[d.name]||'__SECONDARY__').attr('font-size',9)
      .attr('font-family', "'Roboto Mono',monospace").attr('font-weight',500)
      .text(`${(pf_EU[d.name]||0).toLocaleString()} entry`);
    return;
  }
  if(d.step===1){
    const thisU = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step)]||0;
    const step1Total = pf_ST["0"]||1;
    const share1Pct = (thisU/step1Total*100).toFixed(1);
    ll.append('text').attr('x',xpos).attr('y',cy+10).attr('text-anchor',an)
      .attr('fill','__GOOD__').attr('font-size',9)
      .attr('font-family', "'Roboto Mono',monospace").attr('font-weight',600)
      .text(`share 1th step ${share1Pct}%`);
  }
});

const pf_tt=d3.select('#pf-tt'),pf_fmt=n=>n.toLocaleString();
const pf_sT=(h,e)=>pf_tt.html(h).style('opacity',1).style('left',(e.clientX+14)+'px').style('top',(e.clientY-10)+'px');
const pf_hT=()=>pf_tt.style('opacity',0);
const pf_mT=e=>pf_tt.style('left',(e.clientX+14)+'px').style('top',(e.clientY-10)+'px');

pf_ls.on('mouseover',function(ev,d){
  pf_ls.selectAll('path').attr('opacity',.05);d3.select(this).select('path').attr('opacity',.75);
  const srcTot   = (d.source.sourceLinks||[]).reduce((s,l)=>s+l.value,0)||(pf_EU[d.source.name]||1);
  const prevPct  = (d.value/srcTot*100).toFixed(1);
  const entryTot = pf_EU[d.source.name]||srcTot;
  const totalPct = (d.value/entryTot*100).toFixed(1);
  const stepTot  = d.source.step===0?(pf_EU[d.source.name]||1):(pf_ST[String(d.source.step)]||1);
  const stepPct  = (d.value/stepTot*100).toFixed(1);
  pf_sT(`
    <div class="pf-ttt">${d.source.name} \\u2192 ${d.target.name}</div>
    <div class="pf-ttr"><span>Departed \\u2192 Arriving</span><span>${pf_fmt(d.value)}</span></div>
    <div class="pf-ttr"><span>Preference Rate prev step %</span><span>${prevPct}%</span></div>
    <div class="pf-ttr"><span>Total Preference Rate %</span><span>${totalPct}%</span></div>
    <div class="pf-ttr"><span>% of step total</span><span>${stepPct}%</span></div>
    <div class="pf-tts">\\u2191 SOURCE (step ${d.source.step+1})</div>
    <div class="pf-ttr"><span>${d.source.name} sessions</span><span>${pf_fmt(srcTot)}</span></div>
  `,ev);
}).on('mousemove',function(ev){pf_mT(ev);}).on('mouseout',function(){pf_ls.selectAll('path').attr('opacity',.28);pf_hT();});

pf_ns.on('mouseover',function(ev,d){
  const cL=new Set(),cN=new Set([d.id]);
  (d.sourceLinks||[]).forEach(l=>{cL.add(l.index);cN.add(l.target.id);});
  (d.targetLinks||[]).forEach(l=>{cL.add(l.index);cN.add(l.source.id);});
  pf_ls.selectAll('path').attr('opacity',(ld,i)=>cL.has(i)?.68:.04);
  pf_ns.selectAll('rect:first-child').attr('opacity',nd=>cN.has(nd.id)?.95:.18);
  const inc     = (d.targetLinks||[]).reduce((s,l)=>s+l.value,0);
  const out     = (d.sourceLinks||[]).reduce((s,l)=>s+l.value,0);
  const stepTot = d.step===0?(pf_EU[d.name]||1):(pf_ST[String(d.step)]||1);
  const prevSame     = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step-1)];
  const thisSame     = pf_MPS[d.name]&&pf_MPS[d.name][String(d.step)]||inc;
  const nextSameLink = (d.sourceLinks||[]).find(l=>l.target.name===d.name);
  const nextSame     = nextSameLink?nextSameLink.value:0;
  const prefPrevPct  = prevSame&&prevSame>0?(thisSame/prevSame*100).toFixed(1):null;
  const sortedOut = [...(d.sourceLinks||[])].sort((a,b)=>b.value-a.value).slice(0,5);
  const nextHtml  = out>0?sortedOut.map(l=>
    `<div class="pf-ttr"><span>\\u2192 ${l.target.name}</span><span>${pf_fmt(l.value)} (${(l.value/out*100).toFixed(1)}%)</span></div>`
  ).join(''):'';
  const tgtInc  = (d.targetLinks||[]).reduce((s,l)=>s+l.value,0);
  const sortedSrc = [...(d.targetLinks||[])].sort((a,b)=>b.value-a.value).slice(0,3);
  const srcHtml = tgtInc>0?sortedSrc.map(l=>
    `<div class="pf-ttr"><span>\\u2190 ${l.source.name}</span><span>${pf_fmt(l.value)} (${(l.value/tgtInc*100).toFixed(1)}%)</span></div>`
  ).join(''):'';
  if(d.step===0){
    pf_sT(`
      <div class="pf-ttt">${d.name} \\u00b7 Step 1</div>
      <div class="pf-ttr"><span>Entry sessions</span><span>${pf_fmt(pf_EU[d.name]||0)}</span></div>
      <div class="pf-ttr"><span>With next step</span><span>${pf_fmt(out)}</span></div>
      ${nextHtml?`<div class="pf-tts">\\u2192 Next</div>${nextHtml}`:''}
    `,ev);
    return;
  }
  pf_sT(`
    <div class="pf-ttt">${d.name} \\u00b7 Step ${d.step+1}</div>
    ${prevSame?`<div class="pf-ttr"><span>${d.name} at prev step</span><span>${pf_fmt(prevSame)}</span></div>`:''}
    <div class="pf-ttr"><span>${d.name} at this step</span><span>${pf_fmt(thisSame)}</span></div>
    ${prefPrevPct?`<div class="pf-ttr"><span>Preference Rate prev step %</span><span>${prefPrevPct}%</span></div>`:''}
    ${nextSame>0?`<div class="pf-ttr"><span>${d.name} at next step</span><span>${pf_fmt(nextSame)}</span></div>`:''}
    <div class="pf-ttr"><span>Share of step</span><span>${(Math.max(inc,out)/stepTot*100).toFixed(1)}%</span></div>
    ${nextHtml?`<div class="pf-tts">\\u2192 Next</div>${nextHtml}`:''}
    ${srcHtml?`<div class="pf-tts">\\u2190 Top sources</div>${srcHtml}`:''}
  `,ev);
}).on('mousemove',function(ev){pf_mT(ev);})
  .on('mouseout',function(){pf_ls.selectAll('path').attr('opacity',.28);pf_ns.selectAll('rect:first-child').attr('opacity',.88);pf_hT();});

const pf_le=document.getElementById('pf-legend'),pf_un=new Set(pf_gr.nodes.map(n=>n.name));
Object.entries(pf_CL).filter(([n])=>pf_un.has(n)).forEach(([n,c])=>{
  const it=document.createElement('div');it.className='pf-li';
  it.innerHTML=`<div class="pf-ld" style="background:${c}"></div>${n}`;pf_le.appendChild(it);
});

pf_ls.selectAll('path')
  .attr('stroke-dasharray',function(){return this.getTotalLength()+' '+this.getTotalLength();})
  .attr('stroke-dashoffset',function(){return this.getTotalLength();})
  .transition().duration(1400).delay((d,i)=>50+i*4).ease(d3.easeCubicOut)
  .attr('stroke-dashoffset',0).on('end',function(){d3.select(this).attr('stroke-dasharray',null);});
// Save full widget as SVG (visual 1:1 using raster snapshot embedded in SVG)
const pfSaveBtn = document.getElementById('pf-save-svg');
const pfSaveStatus = document.getElementById('pf-save-status');
if (pfSaveBtn){
  pfSaveBtn.addEventListener('click', async () => {
    try {
      const root = document.querySelector('.pfw');
      if (!root) throw new Error('Widget root not found');
      if (typeof html2canvas === 'undefined') throw new Error('html2canvas is not loaded');

      const saveBlock = root.querySelector('.pf-save-bottom');
      const prevDisplay = saveBlock ? saveBlock.style.display : '';
      if (saveBlock) saveBlock.style.display = 'none';
      const canvas = await html2canvas(root, {
        backgroundColor: '#ffffff',
        useCORS: true,
        scale: 2,
      });
      if (saveBlock) saveBlock.style.display = prevDisplay;

      const pngDataUrl = canvas.toDataURL('image/png');
      const width = canvas.width;
      const height = canvas.height;
      const svgString = `<?xml version="1.0" encoding="UTF-8"?>
`
        + `<svg xmlns="http://www.w3.org/2000/svg" width="${width}" height="${height}" viewBox="0 0 ${width} ${height}">`
        + `<rect x="0" y="0" width="${width}" height="${height}" fill="white"/>`
        + `<image href="${pngDataUrl}" x="0" y="0" width="${width}" height="${height}"/>`
        + `</svg>`;

      const resp = await fetch('/__save_svg', {
        method: 'POST',
        headers: {'Content-Type': 'image/svg+xml; charset=utf-8'},
        body: svgString,
      });
      if (!resp.ok) throw new Error('Save failed');
      const msg = await resp.text();
      pfSaveStatus.textContent = msg || 'Saved';
      pfSaveStatus.style.color = '__GOOD__';
    } catch (e) {
      pfSaveStatus.textContent = 'Save failed';
      pfSaveStatus.style.color = '__BAD__';
    }
  });
}

pf_ns.attr('opacity',0).transition().duration(400).delay(d=>d.step*70+120).attr('opacity',1);
</script></body></html>"""

pf_html = (
    pf_html.replace("__BG__", _bg)
    .replace("__CARD__", _card)
    .replace("__PRIMARY__", _primary)
    .replace("__SECONDARY__", _secondary)
    .replace("__BORDER__", _border)
    .replace("__BORDER_LIGHT__", _border_light)
    .replace("__SUBTLE__", XS.BG_SUBTLE)
    .replace("__S12__", XS.S12)
    .replace("__S7__", XS.S7)
    .replace("__S5__", XS.S5)
    .replace("__ACCENT__", XS.S7)
    .replace("__MUTED__", XS.TEXT_HINT)
    .replace("__GOOD__", XS.GOOD)
    .replace("__BAD__", XS.BAD)
    .replace("__FONT_SANS__", "'Open Sans', 'Roboto', sans-serif")
    .replace("__FONT_SANS2__", "'Open Sans', 'Roboto', sans-serif")
    .replace("__SUBTITLE__", pf_subtitle)
    .replace("__KPI_HTML__", pf_kpi_html)
    .replace("__NODES__", json.dumps(pf_all_nodes))
    .replace("__LINKS__", json.dumps(pf_links_data))
    .replace("__STEP_TOT__", json.dumps({str(k): v for k, v in pf_step_totals.items()}))
    .replace("__STEP_LBL__", json.dumps(pf_step_labels))
    .replace("__STEP_SUB__", json.dumps(pf_step_sub))
    .replace("__COLORS__", json.dumps(PF_PID_COLORS))
    .replace("__MAX_STEPS__", str(PF_MAX_STEPS))
    .replace("__ENTRY_USERS__", pf_entry_users_json)
    .replace("__METHOD_PER_STEP__", pf_mps_json)
)


# =========================
# 9) Confluence export
# =========================

pf_conf = (
    pf_html.replace("background:" + _bg + ";", "background:transparent;")
    .replace("padding:24px 16px 36px 32px;", "padding:8px 4px 16px;")
    .replace("max-width:1700px;", "max-width:100%;")
)

pf_export = (
    '<div style="font-family:sans-serif;margin-top:20px;border-radius:12px;'
    'border:1px solid #e2e8f0;overflow:hidden;box-shadow:0 1px 6px rgba(0,0,0,.07);">'
    '<button onclick="var b=document.getElementById(\'pfcb\');var a=document.getElementById(\'pfca\');'
    'var o=b.style.display!==\'none\';b.style.display=o?\'none\':\'block\';'
    'a.style.transform=o?\'rotate(0deg)\':\'rotate(180deg)\';"'
    ' style="width:100%;background:#f8fafc;border:none;border-bottom:1px solid #e2e8f0;'
    'padding:14px 20px;display:flex;align-items:center;justify-content:space-between;cursor:pointer;">'
    '<div><div style="font-size:13px;font-weight:600;color:#0f172a;">Confluence HTML'
    '<span style="background:#EBFCFF;color:#1B8398;font-size:10px;font-weight:600;'
    'padding:2px 8px;border-radius:99px;margin-left:6px;">READY TO PASTE</span></div>'
    '<div style="font-size:11px;color:#94a3b8;margin-top:1px;">'
    'Copy and paste into HTML Macro (height: 1000px)</div></div>'
    '<span id="pfca" style="transition:transform .25s;font-size:18px;color:#94a3b8;">&#8964;</span>'
    '</button>'
    '<div id="pfcb" style="display:none;background:#fff;">'
    '<div style="position:relative;background:#f8fafc;border-bottom:1px solid #f1f5f9;">'
    '<textarea id="pfcc" readonly style="width:100%;height:140px;font-family:monospace;font-size:11px;'
    'color:#475569;background:transparent;border:none;padding:14px 20px;'
    'resize:none;outline:none;line-height:1.6;box-sizing:border-box;">'
    + pf_conf.replace("'", "&#39;")
    + '</textarea>'
    + '<div id="pfck" style="position:absolute;bottom:8px;right:14px;font-size:10px;'
    'color:#cbd5e1;font-family:monospace;pointer-events:none;"></div></div>'
    '<div style="padding:12px 20px;display:flex;align-items:center;gap:12px;">'
    '<button id="pfcpb" onclick="pfCopy()" style="background:#31CAE8;color:#fff;border:none;'
    'border-radius:8px;padding:9px 20px;font-size:13px;font-weight:600;cursor:pointer;transition:background .2s;">'
    'Copy to clipboard</button>'
    '<span id="pfcps" style="font-size:12px;color:#94a3b8;"></span>'
    '</div></div></div>'
    '<script>'
    'var pfta=document.getElementById("pfcc");'
    'document.getElementById("pfck").textContent=(pfta.value.length/1024).toFixed(1)+" KB";'
    'function pfCopy(){'
    'navigator.clipboard.writeText(pfta.value).then(function(){'
    'var btn=document.getElementById("pfcpb"),st=document.getElementById("pfcps");'
    'btn.style.background="#16a34a";btn.textContent="Copied!";'
    'st.textContent="Copied";st.style.color="#16a34a";'
    'setTimeout(function(){btn.style.background="#31CAE8";btn.textContent="Copy to clipboard";st.textContent="";},3000);'
    '}).catch(function(){pfta.select();document.execCommand("copy");});}'
    '</script>'
)

# Jupyter sanitizes inline <script> in HTML(); serve the same file over HTTP and embed via IFrame (D3 works).
import functools
import http.server
import socketserver
import threading
import webbrowser

_pf_preview_path = (Path.cwd() / "payment_flow_sankey_preview.html").resolve()
_pf_preview_path.write_text(pf_html, encoding="utf-8")

class _PfTCPServer(socketserver.TCPServer):
    allow_reuse_address = True


class _PfQuietHandler(http.server.SimpleHTTPRequestHandler):
    def log_message(self, format, *args):
        return

    def do_POST(self):
        if self.path != '/__save_svg':
            self.send_error(404, 'Not Found')
            return
        try:
            content_length = int(self.headers.get('Content-Length', '0'))
            payload = self.rfile.read(content_length)
            target = Path(self.directory) / 'payment_flow_bank_cards_only_successful.svg'
            target.write_bytes(payload)
            self.send_response(200)
            self.send_header('Content-Type', 'text/plain; charset=utf-8')
            self.end_headers()
            self.wfile.write(f'Saved: {target.name}'.encode('utf-8'))
        except Exception as exc:
            self.send_response(500)
            self.send_header('Content-Type', 'text/plain; charset=utf-8')
            self.end_headers()
            self.wfile.write(f'Error: {exc}'.encode('utf-8'))


_pf_root = _pf_preview_path.parent
_pf_httpd = None
_pf_port = None
for _pf_try_port in range(8765, 8795):
    try:
        _pf_handler = functools.partial(
            _PfQuietHandler,
            directory=str(_pf_root),
        )
        _pf_httpd = _PfTCPServer(("127.0.0.1", _pf_try_port), _pf_handler)
        _pf_port = _pf_try_port
        break
    except OSError:
        continue

if _pf_httpd is not None and _pf_port is not None:
    threading.Thread(target=_pf_httpd.serve_forever, daemon=True).start()
    _pf_iframe_url = f"http://127.0.0.1:{_pf_port}/{_pf_preview_path.name}"
    print("Embedded chart (HTTP iframe, scripts allowed):", _pf_iframe_url)
    display(IFrame(src=_pf_iframe_url, width="100%", height=500))
else:
    print("Could not start local HTTP preview; open the HTML file or use the VS Code task.")
































```
